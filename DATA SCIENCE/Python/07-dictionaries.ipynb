### Python Dictionaries

A **Dictionary** is an unordered, mutable collection of **key-value pairs**. It is designed to store data values like a map, unlike sequences (lists, tuples) which store data by numerical index.

Dictionaries are defined using **curly braces** (`{}`) and separate keys from values with a **colon** (`:`) and key-value pairs with a **comma** (`,`).

**Key Characteristics:**

1.  **Unordered** (as of Python 3.7+ they maintain insertion order).
2.  **Mutable:** You can add, remove, and change key-value pairs.
3.  **Keys must be Unique:** No two keys in a dictionary can be the same.
4.  **Keys must be Immutable:** Keys must be of an immutable type (e.g., strings, numbers, tuples), but **not** lists or dictionaries.

```python
# Creating a simple dictionary
person = {
    "name": "Alex",
    "age": 28,
    "city": "London"
}
print(f"Person Dictionary: {person}")
print(f"Type: {type(person)}")

# Keys can be of different immutable types
mixed_keys = {
    1: "Integer Key",
    "two": "String Key",
    (3, 4): "Tuple Key"
}
print(f"Mixed Keys: {mixed_keys}")
```

### Indexing & Slicing

Unlike lists and tuples, dictionaries do not use numerical indices. Data is accessed using its unique **key**.

#### Accessing Values

  * **Using Square Brackets (`[]`):** This is the standard way to access a value. If the key is not found, it raises a **`KeyError`**.
  * **Using the `.get()` method:** This is a safer way. If the key is not found, it returns `None` or a specified default value, rather than raising an error.

```python
student = {"id": 101, "name": "Maria", "grade": "A"}

# Accessing using square brackets
student_name = student["name"]
print(f"Student Name (using []): {student_name}")

# Accessing using .get()
student_grade = student.get("grade")
print(f"Student Grade (using .get()): {student_grade}")

# Handling a missing key with [] (raises an error)
print(student["address"])
```

```python
# Handling a missing key with .get() (returns None or default)
student_address = student.get("address", "N/A")
print(f"Student Address (using .get() with default): {student_address}")
```

#### Slicing

Dictionaries **do not support slicing**. You must iterate through the dictionary or convert it to a sequence (like a list of keys) to extract subsets based on criteria.

### Modifying a Dictionary

Dictionaries are mutable. You can add new key-value pairs, change existing values, or remove pairs.

#### Adding and Updating

  * To **add a new key-value pair**, assign a value to a new key.
  * To **change an existing value**, assign a new value to the existing key.

<!-- end list -->

```python
user = {"username": "coder", "status": "active"}
print(f"Original User: {user}")

# 1. Updating an existing key
user["status"] = "offline"
print(f"After Update: {user}")

# 2. Adding a new key-value pair
user["role"] = "admin"
print(f"After Adding 'role': {user}")
```

```python
user.update({"role": "user", "address": "new york"})
```

#### Removing Items

  * **`del` keyword:** Deletes the key-value pair.
  * **`.pop(key)`:** Removes the item with the specified key and returns the value. Raises a `KeyError` if the key is not found (unless a default is specified).
  * **`.popitem()`:** Removes and returns an **arbitrary** (or the last inserted, in modern Python) key-value pair as a tuple.
  * **`.clear()`:** Empties the entire dictionary.

<!-- end list -->

```python
settings = {"theme": "dark", "size": "medium", "notifications": True}

# Using del
del settings["size"]
print(f"After del 'size': {settings}")

# Using .pop()
removed_value = settings.pop("notifications")
print(f"Pop value: {removed_value}")
print(f"After pop: {settings}")
```

### Copy vs Reference

Just like lists, simple assignment with dictionaries creates a **reference** (or view). To create an independent copy, you must use the `.copy()` method or the `dict()` function.

  * **Reference (View):** Simple assignment (`dict_b = dict_a`). Changing one modifies the other.
  * **Copy:** Use `.copy()` or `dict()`. Changing the copy does **not** affect the original.

<!-- end list -->

```python
original_data = {"a": 1, "b": 2}

# 1. Reference (View)
view_data = original_data
view_data["a"] = 99  # Change view_data
print(f"Original (after view change): {original_data}") # Original IS CHANGED: {'a': 99, 'b': 2}

# 2. Copy
original_data = {"a": 1, "b": 2} # Reset original
copy_data = original_data.copy()
# copy_data = dict(original_data) # Alternative way to copy

copy_data["a"] = 55 # Change copy_data
print(f"Original (after copy change): {original_data}") # Original REMAINS UNCHANGED: {'a': 1, 'b': 2}
```

### Dictionary Methods

Dictionaries come with powerful methods for iterating and managing views of the keys, values, and items.

| Method | Description | Example |
| :--- | :--- | :--- |
| **`.keys()`** | Returns a **view object** of all keys in the dictionary. | `{'a': 1}.keys()` $\rightarrow$ `dict_keys(['a'])` |
| **`.values()`** | Returns a **view object** of all values in the dictionary. | `{'a': 1}.values()` $\rightarrow$ `dict_values([1])` |
| **`.items()`** | Returns a **view object** of all key-value pairs (as tuples). | `{'a': 1}.items()` $\rightarrow$ `dict_items([('a', 1)])` |
| **`.update(iterable)`** | Updates the dictionary with key-value pairs from another dictionary or iterable. | `{'a': 1}.update({'b': 2})` $\rightarrow$ `{'a': 1, 'b': 2}` |
| **`.setdefault(key, value)`** | Returns the value of the specified key. If the key doesn't exist, it inserts the key with the specified value. | `d.setdefault('c', 0)` |

```python
profile = {"user": "jane_d", "last_login": "2024-01-01", "is_active": True}

# Getting keys, values, and items views
print(f"Keys: {profile.keys()}")
print(f"Values: {profile.values()}")
print(f"Items: {profile.items()}")

# Using .update()
new_data = {"is_active": False, "email": "jane@example.com"}
profile.update(new_data)
print(f"\nAfter Update: {profile}")
```

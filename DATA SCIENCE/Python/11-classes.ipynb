# Introduction

Python is primarily and object-oriented programming (OOP) language. In OOP, the world is viewed as a collection of interacting objects. While there are built in objects such as list, tuples, integers in Python, we also create our own objects by defining their class.

A class is a blueprint, a template, or a prototype from which objects are created. It defines a set of attributes (data) and methods (functions) that the objects created from it will possess. Think of a class like the design of a building.

An object (instance of a class) is a concrete realization of a class. It's an individual instance created from the class blueprint, with its own unique data. Following the building design analogy, an object is a house built based on the design.

For example. If `Car` is a class, then `my_car` (a red Toyota Camry) and `your_car` (a blue Honda Civic) are objects of the `Car` class. Both `my_car` and `your_car` have attributes like `color`, `make`, `model`, and methods like `start()`, `stop()`.

## Creating a Class

Classes are defined using the `class` keyword. By convention, class names use `PascalCase` (first letter of each word capitalized).

```python
class Dog:
    # The __init__ method (constructor)
    # self refers to the instance of the class
    def __init__(self, name, age):
        self.name = name  # Instance attribute
        self.age = age    # Instance attribute

    # Instance method
    def bark(self):
        return f"{self.name} barks!"

    # Another instance method
    def description(self):
        return f"{self.name} is {self.age} years old."
```

`class Dog:` defines a class named Dog. `def __init__(self, name, age):` is a special method called the constructor. It's automatically called when a new object is created. `self` parameter refers to the instance of the class itself. It must be the first parameter of any instance method. `self.name = name` is used to set an instance variable called `name`. Each Dog object will have its own unique name which can be accessed via the `name` attribute of objects created with the class. `def bark(self):`: This is an instance method. It operates on the specific instance of the class (`self`).

```python
# Creating objects (instances) of the Dog class
dog1 = Dog("Buddy", 3)
dog2 = Dog("Lucy", 5)

# Accessing object attributes
print(f"Dog 1's name: {dog1.name}")
print(f"Dog 2's age: {dog2.age}")

# Calling object methods
print(dog1.bark())
print(dog2.description())
```

## Class and Instance Variables

Class variables are declared directly inside the class definition, but outside any method. They are shared by all instances (objects) of the class, accessed using either `ClassName.variable_name` or `object_name.variable_name`. If accessed via `object_name.variable_name`, and the instance doesn't have an attribute with that name, Python looks for it in the class.They are useful for data that is common to all instances.

Instance variables are declared inside a method (typically `__init__`) using `self.variable_name`. They are unique to each instance (object) of the class. Each object can have different values for its instance variables. They are accessed using `object_name.variable_name`.

Let's took at an example:

```python
class Car:
    # Class variable
    wheels = 4

    def __init__(self, make, model, color):
        # Instance variables
        self.make = make
        self.model = model
        self.color = color

car1 = Car("Toyota", "Camry", "Red")
car2 = Car("Honda", "Civic", "Blue")

print(f"Car 1 make: {car1.make}") # Accessing instance variable
print(f"Car 2 color: {car2.color}") # Accessing instance variable

print(f"Car 1 wheels: {car1.wheels}") # Accessing class variable via instance
print(f"Car 2 wheels: {car2.wheels}")
print(f"Car class wheels: {Car.wheels}") # Accessing class variable via class

# Changing class variable affects all instances (if accessed via class or inherited)
Car.wheels = 5
print("After changing Car.wheels:")
print(f"Car 1 wheels: {car1.wheels}")
print(f"Car 2 wheels: {car2.wheels}")
```

We can also modify instance variables by assigning new values to attribute of the objects.

```python
car1.color = "Green"
car2.color = "Grey"

print(f"Car 1 new color: {car1.color}")
print(f"Car 2 new color: {car2.color}")
```

## Inheritance

Inheritance is a mechanism that allows a new class (subclass/child class) to inherit properties and behaviors from an existing class (superclass/parent class). This promotes code reusability and establishes an "is-a" relationship (e.g., a `Dog` "is a" `Animal`).

```python
# Parent class
class Animal:
    def __init__(self, name, sound):
        self.name = name
        self.sound = sound

    def make_sound(self):
        return f"{self.name} says {self.sound}"

    def describe(self):
        return f"This is an animal named {self.name}."

# Child class inheriting from Animal
class Dog(Animal):
    def __init__(self, name, breed):
        # Call the parent class's constructor
        super().__init__(name, "Woof!") # Dog's sound is always "Woof!"
        self.breed = breed

    # Override parent method
    def describe(self):
        return f"{self.name} is a {self.breed} dog."

    def fetch(self):
        return f"{self.name} fetches the ball!"

# Another child class
class Cat(Animal):
    def __init__(self, name, color):
        super().__init__(name, "Meow!")
        self.color = color

```

Dog inherits from Animal. Animal is the parent, Dog is the child. The super() function is used to call methods of the parent class. Here, it calls the `__init__` of Animal to initialize name and sound. When a child class defines a method with the same name as a method in its parent class (e.g., `describe()` in `Dog`), the child's version of the method is used.

```python
# Creating objects
my_dog = Dog("Rex", "German Shepherd")
my_cat = Cat("Whiskers", "Orange")

print(my_dog.make_sound()) # Inherited method
print(my_dog.describe()) # Overridden method
print(my_dog.fetch()) # Child class specific method

print(my_cat.make_sound())
print(my_cat.describe())
print(f"{my_cat.name} is {my_cat.color}.")
```

## Dunder Methods (Magic Methods)

Dunder methods (short for "double underscore" methods) are special methods in Python that have leading and trailing double underscores (e.g., `__init__`, `__str__`, `__len__`). They allow you to define how objects of your class behave with built-in functions and operators.

Some common dunder methods:

* `__init__`(self, ...): Constructor, called when an object is created.

* `__str__`(self): Defines the informal string representation of an object (what print() shows).

* `__repr__`(self): Defines the official string representation (used for debugging, typically shows how to recreate the object).

* `__len__`(self): Defines the behavior for the len() function.

* `__eq__`(self, other): Defines the behavior for the == operator.

* `__add__`(self, other): Defines the behavior for the + operator.

```python
class Book:
    def __init__(self, title, author, pages):
        self.title = title
        self.author = author
        self.pages = pages

    # __str__ for user-friendly output
    def __str__(self):
        return f"'{self.title}' by {self.author}"

    # __repr__ for unambiguous representation
    def __repr__(self):
        return f"Book('{self.title}', '{self.author}', {self.pages})"

    # __len__ to allow len() on Book objects
    def __len__(self):
        return self.pages

    # __eq__ to compare books by title and author
    def __eq__(self, other):
        if isinstance(other, Book):
            return self.title == other.title and self.author == other.author
        return False

book1 = Book("The Hitchhiker's Guide to the Galaxy", "Douglas Adams", 193)
book2 = Book("Pride and Prejudice", "Jane Austen", 279)
book3 = Book("The Hitchhiker's Guide to the Galaxy", "Douglas Adams", 200)

print(book1) # Uses __str__
print(repr(book1)) # Uses __repr__

print(f"Number of pages in book1: {len(book1)}") # Uses __len__

print(f"book1 == book2: {book1 == book2}") # Uses __eq__
print(f"book1 == book3: {book1 == book3}")
```

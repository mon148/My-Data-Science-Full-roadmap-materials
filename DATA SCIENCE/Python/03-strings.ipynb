##  Python Strings

### Python Strings

A **string** in Python is a sequence of characters, used to store text. Strings are **immutable**, meaning once they are created, they cannot be changed.

Strings can be created by enclosing characters in either **single quotes** (`'...'`) or **double quotes** (`"..."`).

```python
# Single quotes
my_string_1 = 'Hello, World!'
print(my_string_1)
print(type(my_string_1))

# Double quotes
my_string_2 = "Python is fun."
print(my_string_2)

# Multi-line strings can be created using three quotes (single or double)
multi_line_string = """This is a string
that spans across
multiple lines."""
print(multi_line_string)
```

### Slicing Strings

**Indexing** allows you to access individual characters. Python uses **zero-based indexing**, meaning the first character is at index 0. **Negative indexing** accesses characters from the end of the string, with $-1$ being the last character.

**Slicing** allows you to get a substring (a range of characters) using the slice syntax: `[start:stop:step]`. The `stop` index is **exclusive**.

```python
my_string = "ABCDEFGHIJKLMNOP"

# Accessing a single character (Indexing)
print("Character at index 0:", my_string[0])
print("Character at index 5:", my_string[5])

# Accessing the last character (Negative Indexing)
print("Last character:", my_string[-1])
print("Third to last character:", my_string[-3])

# Slicing from index 2 up to (but not including) index 5
print("Slice [2:5]:", my_string[2:5]) # Output: CDE

# Slicing from the beginning up to index 4
print("Slice [:4]:", my_string[:4]) # Output: ABCD

# Slicing from index 7 to the end
print("Slice [7:]:", my_string[7:]) # Output: HIJKLMNOP

# Slicing with a step (every 2nd character)
print("Slice [::2]:", my_string[::2]) # Output: ACEGIKMO

# Reversing a string
print("Reversed string:", my_string[::-1]) # Output: PONMLKJIHGFEDCBA
```

### Modifying Strings

Since strings are immutable, you **cannot** change a character in place. Any "modification" operation actually creates a **new string**.

```python
original_string = "apple"

# Attempting to change a character will result in a TypeError
original_string[0] = 'A' # Expected output: 'Apple'
```

```python
# To 'modify' a string, you must create a new one, often using slicing or methods
new_string = 'A' + original_string[1:]
print("New string:", new_string) # Output: Apple
```

### Concatenating Strings

**Concatenation** is the process of joining two or more strings together, typically using the `+` operator.

```python
str_a = "Hello"
str_b = " "
str_c = "World"
str_d = "!"

full_greeting = str_a + str_b + str_c + str_d
print(full_greeting)

# You can also use the '*' operator to repeat a string
repeated_string = "ha" * 3
print(repeated_string)
```

### Escape Characters

To insert characters that are illegal in a string (like a double quote inside a double-quoted string), you use an **escape character**, which is a **backslash** (`\`).

Common escape characters:

  * `\'`: Single Quote
  * `\"`: Double Quote
  * `\\`: Backslash
  * `\n`: New Line
  * `\t`: Tab

<!-- end list -->

```python
# Using a double quote inside a double-quoted string
text_1 = "He said, \"I'm hungry.\""
print(text_1)

# Inserting a newline
text_2 = "Line one.\nLine two."
print(text_2)

# Inserting a tab
text_3 = "Item 1\t$10"
print(text_3)

# Inserting a backslash
text_4 = "The path is C:\\Users\\Name"
print(text_4)
```

### Formatted Strings

Python offers several ways to format strings. The most modern and preferred method is using **f-strings** (formatted string literals), introduced in Python 3.6. They allow you to embed expressions inside string literals using curly braces (`{}`).

```python
name = "Alice"
age = 30
pi_value = 3.14159

# Using f-strings (f-string literals)
message_f = f"Hello, my name is {name} and I am {age} years old."
print(message_f)

# F-strings support expressions and formatting specifiers
area = 5 * 5
pi_formatted = f"The area of a 5x5 square is {area}. Pi is approximately {pi_value:.2f}."
print(pi_formatted)

# Older methods (kept for reference, but less common now)
# The .format() method
message_format = "Hello, my name is {} and I am {} years old.".format(name, age)
print(f"Using .format(): {message_format}")

# The old %-formatting (C-style)
message_percent = "Hello, my name is %s and I am %d years old." % (name, age)
print(f"Using %-formatting: {message_percent}")
```

### String Methods

Python provides a rich set of built-in methods for strings that return new strings. Remember, they don't modify the original string.

| Method | Description | Example |
| :--- | :--- | :--- |
| **`.upper()`** | Converts a string to uppercase. | `"hello".upper()` $\rightarrow$ `"HELLO"` |
| **`.lower()`** | Converts a string to lowercase. | `"WORLD".lower()` $\rightarrow$ `"world"` |
| **`.strip()`** | Removes leading and trailing whitespace. | `"  text  ".strip()` $\rightarrow$ `"text"` |
| **`.replace(old, new)`**| Replaces all occurrences of a substring. | `"bad day".replace("bad", "good")` $\rightarrow$ `"good day"` |
| **`.split(separator)`** | Splits the string into a list of substrings. | `"a,b,c".split(',')` $\rightarrow$ `['a', 'b', 'c']` |
| **`.find(substring)`** | Searches for a substring and returns the first index. Returns $-1$ if not found. | `"apple".find("p")` $\rightarrow$ $1$ |
| **`.count(substring)`** | Returns the number of times a specified value occurs. | `"banana".count("a")` $\rightarrow$ $3$ |

```python
sample_text = "  Python Programming  "

# Using upper() and lower()
print(f"Original: '{sample_text}'")
print(f"Uppercase: '{sample_text.upper()}'")
print(f"Lowercase: '{sample_text.lower()}'")

# Using strip() to remove surrounding whitespace
stripped_text = sample_text.strip()
print(f"Stripped: '{stripped_text}'")

# Using replace()
replaced_text = stripped_text.replace("Programming", "Language")
print(f"Replaced: '{replaced_text}'")

# Using split()
words = replaced_text.split(" ")
print(f"Split into a list: {words}")

# Checking properties
print(f"'123'.isdigit(): {'123'.isdigit()}") # True
print(f"'hello'.isalpha(): {'hello'.isalpha()}") # True
```

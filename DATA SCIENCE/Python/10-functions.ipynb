## Functions

Functions are reusable blocks of code that perform a specific task. They help organize code, make it more readable, and avoid repetition.

* **Defining a Function:** Use the `def` keyword followed by the function name, parentheses `()`, and a colon `:`. Parameters (inputs) can be specified within the parentheses. The code block within the function is indented.

* **Calling a Function:** To execute a function, use its name followed by parentheses `()`. If the function has parameters, provide the arguments (values) within the parentheses.

* **`return` Statement:** Used to send a value back from the function to the caller. If no `return` statement is present, the function returns `None`.

### Functions without arguments (parameters)

A function can be defined without any parameters. When called, it simply executes the code within its block.

````python
def say_hello():
    """This function prints a greeting."""
    print("Hello!")
````

````python
say_hello()
````

### Functions with positional arguments (parameters)
When defining a function with parameters, the arguments passed during the function call are matched to the parameters based on their order or position.

````python
def greet(name): # A function with only one paremeter
    """This function greets the person passed in as a parameter."""
    print(f"Hello, {name}!")

greet("Alice")  # Calling the function. The string "Alice" is assigned to the parameter (variable) name.
````

````python
def describe_pet(animal_type, pet_name): # A function with two parameters
    """This function describes a pet."""
    print(f"I have a {animal_type} named {pet_name}.")

describe_pet("dog", "Buddy")      # Positional arguments - "dog" is assigned to animal_type, "Buddy" to pet_name
describe_pet("Whiskers", "cat")   # Incorrect order leads to unintended output
````

### Functions with keyword arguments
To avoid confusion with positional arguments and to make function calls more explicit, you can use keyword arguments. When calling a function with keyword arguments, you specify the parameter name followed by the value using the = sign. The order of keyword arguments doesn't matter.

````python
def describe_pet(animal_type, pet_name):
    """This function describes a pet."""
    print(f"I have a {animal_type} named {pet_name}.")

describe_pet(animal_type="dog", pet_name="Pip")  # Keyword arguments
describe_pet(pet_name="Lucy", animal_type="goldfish") # Order doesn't matter with keywords
````

### Functions with default arguments
When defining a function, we can provide default values for parameters. If a value is not provided for that parameter during the function call, the default value is used. Default arguments make function calls more flexible.

````python
def power(base, exponent=2):
    """Calculates the power of a number (default exponent is 2)."""
    result = base ** exponent
    return result

print(power(5))       # Uses the default exponent (2) - Output: 25
print(power(5, 3))    # Overrides the default exponent with 3 - Output: 125
print(power(base=10)) # Uses the default exponent (2) with a keyword argument - Output: 100
````

### Collecting and unpacking parameters in function definition and function calls

Python's `*` (asterisk) and `**` (double asterisk) operators are used for "unpacking" iterables and dictionaries, respectively, primarily within function definitions and function calls. This allows for highly flexible argument handling.

#### Collecting and unpacking positinal arguments

The `*` operator is used to unpack an iterable (like a list or tuple) into individual positional arguments.

**In function definition (collecting positonal arguments)**:

When `*args` is used as a parameter in a function definition, it collects an arbitrary number of positional arguments passed to the function into a tuple. The name `args` is a convention; you could use any valid variable name.

For example:

````python
def my_func(*arg):
    print(arg)
````

````python
my_func("a", "b", "c", "d")
````

````python
def my_func2(arg1, *arg2):
    print("Argument 1:", arg1)
    print("Argument 2:", arg2)
````

````python
my_func2("a", "b", "c", "d")
````

**In function calls (unpacking iterables):**

When `*iterable` is used in a function call, it unpacks the elements of iterable (e.g., a list or tuple) and passes them as individual positional arguments to the function.

For example:

````python
def multiply(a, b, c):
    return a * b * c
````

````python
my_numbers = [2, 3, 4]
result = multiply(*my_numbers) # Equivalent to multiply(2, 3, 4)
result
````

#### Collecting and unpacking keyword arguments


The `**` operator is used to collect keyword arguments into a dictionary or unpack a dictionary into individual keyword arguments.

**In function definition (collecting keyword arguments):

When `**kwargs` is used as a parameter in a function definition, it collects an arbitrary number of keyword arguments passed to the function into a dictionary. The name `kwargs` is a convention, you name the parameter anything.

````python
def my_func(**kwargs):
    print(kwargs)
````

````python
my_func(a=1, b=2, c=3)
````

````python
def describe_person(name, **details):
    print(f"Name: {name}")
    for key, value in details.items():
        print(f"{key.replace('_', ' ').title()}: {value.replace('_', ' ')}")
````

````python
describe_person("Bobby", job_role="ML_Engineer", nationality="Nigerian")
````

**In function calls (unpacking dictionaries)**:

When `**my_dict` is used in a function call, it unpacks the key-value pairs of my_dict and passes them as individual keyword arguments to the function.

````python
def configure_settings(theme, font_size, language):
    print(f"Theme: {theme}")
    print(f"Font Size: {font_size}")
    print(f"Language: {language}")

my_config = {"theme": "dark", "font_size": 14, "language": "fr"}
configure_settings(**my_config) # Equivalent to configure_settings(theme="dark", font_size=14, language="fr")
````

### Combining `*args` and `**kwargs`

You can use both `*args` and `**kwargs` in the same function definition. When doing so, `*args` must come before `**kwargs`.

````python
def my_func(arg, *args, **kwargs):
    print("First argument:", arg)
    print("Second argument(s):", args)
    print("Keyword arguments:", kwargs)

my_func("a", "b", "c", d="d", e="e")
````

### Variable scopes

In Python, the scope of a variable determines where in your code that variable can be accessed and modified. Understanding this is crucial for working with functions.

#### Local variables

A local variable is a variable that is defined inside a function. It is only accessible from within that function. The key characteristics include:

* **Scope:** They exist only within the function where they are defined.
* **Lifetime:** They are created when the function is called and destroyed once the function finishes execution.
* **Isolation:** Changes made to a local variable within a function do not affect any other variable outside that function, even if another variable has the same name.
* **Reusability of names:** You can use the same variable name in different functions without them interfering with each other. Each instance will be a separate local variable.

````python
def my_function():
    # 'x' is a local variable
    x = 10
    print(f"Inside the function, x is: {x}")

my_function()
````

````python
print(x)
````

### Global Variables

A **global variable** is a variable that is defined *outside* of any function, typically at the top level of a script. It is accessible from anywhere in the code, both inside and outside functions. The key characteristics include:

* **Scope:** They are accessible throughout the entire program.
* **Lifetime:** They are created when the program starts and destroyed when the program ends.
* **Read access:** Functions can read the value of a global variable without any special declaration.
* **Write access (with `global` keyword):** To *modify* a global variable inside a function, you must explicitly declare it as `global` within the function using the `global` keyword. If you don't, Python will create a new local variable with the same name, rather than modifying the global one.

````python
# 'y' is a global variable
y = 20

def another_function():
    print(f"Inside the function, y is: {y}") # We can read 'y'

another_function()
print(f"Outside the function, y is: {y}")
````

Modifying a global variable using `global`:

````python
z = 30

def modify_global():
    global z  # Declare 'z' as global to modify the global variable
    z = 40
    print(f"Inside the function, z is: {z}")

print(f"Before calling function, z is: {z}")
modify_global()
print(f"After calling function, z is: {z}")
````

*Not* modifying a global variable (creates a local one instead):

````python
# 'a' is a global variable
a = 50

def attempt_to_modify():
    a = 60  # This creates a NEW LOCAL variable named 'a', not modifying the global 'a'
    print(f"Inside the function, local 'a' is: {a}")

print(f"Before calling function, global 'a' is: {a}")
attempt_to_modify()
print(f"After calling function, global 'a' is: {a}") # Global 'a' remains unchanged
````

## Lambda Functions (Anonymous Functions)

In addition to defining functions using the `def` keyword, Python also allows you to create small, anonymous functions using the lambda keyword. These are often referred to as lambda functions.

```
lambda arguments: expression
```
Lambda functions are frequently used in situations where a small function is needed for a short period, often as arguments to other functions (like `map()`, `filter()`, `sorted()`).

````python
# A lambda function to add two numbers

add = lambda x, y: x + y
result = add(5, 3)
print(result)  # Output: 8
````

````python
# Using a lambda function with map() to square each element in a list

numbers = [1, 2, 3, 4, 5]
squares = list(map(lambda x: x ** 2, numbers))
print(squares)
````

````python
# Using a lambda function with filter() to get even numbers from a list

numbers = [1, 2, 3, 4, 5, 6]
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
print(even_numbers)
````

````python
# Using a lambda function as the key for sorting a list of tuples

data = [("Alice", 25), ("Bob", 30), ("Charlie", 22)]
sorted_data = sorted(data, key=lambda item: item[1])  # Sort by age (second element)
print(sorted_data)
````

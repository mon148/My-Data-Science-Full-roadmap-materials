A regular expression (often abbreviated as regex or regexp) is a special sequence of characters that defines a search pattern. Regular expressions are used for matching, searching, and manipulating text based on specific patterns or rules.

They are widely used in programming, data processing, text editing, and data validation tasks-such as finding email addresses in a document, validating phone numbers, or extracting specific information from text. Most programming languages (like Python, Java, JavaScript, and Perl) and many text editors support regular expressions.

You can go to a website such as [Regex Pal](https://www.regexpal.com) and test your regular expressions before using them your code. To work with regular expressions in python we use the "re" module.

```python
import re
```

## Basic Matching with `re.search()`

The `re.search()` function is your go-to for checking if a pattern exists anywhere in a string. It returns a match object if the pattern is found, and None otherwise.

```python
text = "The quick brown fox jumps over the lazy dog."
pattern = "fox"

match = re.search(pattern, text)

if match:
    print("Found a match!")
    print("Starting position:", match.start())
    print("Ending position:", match.end())
    print("Matched string:", match.group())
else:
    print("No match found.")
```

## Common re Module Functions

Besides `re.search()`, here are some other useful functions:

* **re.findall(pattern, string)**: Finds all non-overlapping matches of the pattern in the string and returns them as a list of strings.

* **re.match(pattern, string)**: Attempts to match the pattern at the beginning of the string. Returns a match object if successful, None otherwise.

* **re.sub(pattern, replacement, string)**: Replaces all occurrences of the pattern in the string with the replacement string.

## Special Characters (Metacharacters)

The real power of regex comes from its special characters that represent more than just themselves. Here are a few fundamental ones:

### `.` (dot): Matches any single character (except newline).

```python
text = "cat bat mat hat bot attraction boat"
pattern = ".at"  # Matches any character followed by "at"
matches = re.findall(pattern, text)
print(matches)
```

### `^` (caret): Matches the beginning of the string.

```python
text = "hello world"
pattern = "^hello"  # Matches if the string starts with "hello"
match = re.search(pattern, text)
if match:
    print("Starts with 'hello'")
```

### `$` (dollar sign): Matches the end of the string.

```python
text = "hello world"
pattern = "world$"  # Matches if the string ends with "world"
match = re.search(pattern, text)
if match:
    print("Ends with 'world'")
```

### `*` (asterisk): Matches the preceding character zero or more times.

```python
text = "b bo boo booo"# Output: ['b', 'bo', 'boo', 'booo']
pattern = "oo*"  # Matches "b" followed by zero or more "o"s
matches = re.findall(pattern, text)
print(matches)
```

### `+` (plus sign): Matches the preceding character one or more times.

```python
text = "b bo boo booo"
pattern = "bo+"  # Matches "b" followed by one or more "o"s
matches = re.findall(pattern, text)
print(matches)  # Output: ['bo', 'boo', 'booo']
```

### `?` (question mark): Matches the preceding character zero or one time (optional).

```python
text = "color colour column"
pattern = "colou?r"  # Matches "color" or "colour"
matches = re.findall(pattern, text)
print(matches)
```

### `[]` (square brackets)
This defines a character set. It matches any single character within the brackets. Inserting `^` at the begining of the character set (`[^]`) matches any single character not within the brackets.

```python
text = "cat bat mat 1at"
pattern = "[cbm]at"  # Matches "cat", "bat", or "mat"
matches = re.findall(pattern, text)
print(matches)

pattern = "[a-z]at"  # Matches any lowercase letter followed by "at"
matches = re.findall(pattern, text)
print(matches)

pattern = "[0-9]at"  # Matches any digit followed by "at"
matches = re.findall(pattern, text)
print(matches)

pattern = "[^c]at"  # Matches any character EXCEPT "c" followed by "at"
matches = re.findall(pattern, text)
print(matches)
```

```python
text = "cat boul bat mat 1at foul"
re.findall("[a-z]+l", text)
```

### `\` (backslash)
It's used to escape special characters or indicate special sequences. For example, \. matches a literal dot, \d matches any digit, \s matches any whitespace character.

It's highly recommended to use raw strings (prefixing the string with r) when defining regular expression patterns. This prevents Python from interpreting backslashes as escape sequences, which can be crucial in regex where backslashes have special meaning.

```python
text = "Price: $10.99"
pattern = r"\$[0-9]+\.[0-9]+"  # Matches a dollar sign, one or more digits, a dot, and one or more digits
match = re.search(pattern, text)
if match:
    print("Found price:", match.group())

text = "word1 word2\tword3\nword4"
pattern = r"\s+"  # Matches one or more whitespace characters (space, tab, newline)
matches = re.split(pattern, text)
print(matches)
```

### `()` (parentheses)
Creates a capturing group. It groups the pattern inside, and you can extract the matched parts.

```python
text = "user@example.com"
# pattern = r"(\w+)@(\w+)\.(\w+)"  # Captures username, domain, and top-level domain
pattern = r"([a-z]+)@([a-z]+)\.([a-z]+)"
match = re.search(pattern, text)
if match:
    print("Username:", match.group(1))
    print("Domain:", match.group(2))
    print("Top-level domain:", match.group(3))
```

Some other commonly used regular expression symbols are described in the table below:

| Symbol | Meaning                                                                    | Example Pattern | Example String        | Example Matches      |
| :----- | :------------------------------------------------------------------------- | :-------------- | :-------------------- | :------------------- |
| `{n}`  | Matches the preceding character exactly `n` times.                         | `a{3}`          | `"baaaac"`            | `aaa`                |
| `{n,}` | Matches the preceding character `n` or more times.                         | `go{2,}`        | `"google goooogle"`   | `goo`, `goooo`   |
| `{n,m}`| Matches the preceding character between `n` and `m` times (inclusive).      | `[0-9]{2,4}`    | `"Year: 2023, 99, 1985"` | `2023`, `99`, `1985` |
| `\w`   | Matches any word character (alphanumeric and underscore: `a-zA-Z0-9_`).    | `\w+`           | `"hello_world123"`    | `hello_world123`     |
| `\W`   | Matches any non-word character (anything not alphanumeric or underscore). | `\W+`           | `"space!@#$"`         | ` !@#$`              |
| `\d`   | Matches any digit (0-9).                                                   | `\d{3}-\d{3}-\d{4}` | `"123-456-7890"`      | `123-456-7890`       |
| `\D`   | Matches any non-digit character.                                           | `\D+`           | `"Letters and spaces"` | `Letters and spaces` |
| `\s`   | Matches any whitespace character (space, tab, newline, etc.).             | `word\s+word`   | `"word   another word"` | `word   another word` |
| `\S`   | Matches any non-whitespace character.                                        | `\S+`           | `"non white space"`   | `non`, `white`, `space` |
| `\|`    | Matches any character, string of characters, or subexpression, separated by the "\|" | `b(a\|i\|e)d` | `"bad bid bed"`   | `bad`, `bid`, `bed` |

These symbols can be combined to create even more intricate and powerful search patterns.

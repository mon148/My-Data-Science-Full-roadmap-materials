### Python List

A **List** in Python is an ordered, mutable sequence of items. It is one of the most flexible built-in data types, as items in a list do not have to be of the same type.

Lists are defined by enclosing comma-separated items within **square brackets** (`[]`).

```python
# A list of integers
numbers = [1, 2, 3, 4, 5]
print(f"Numbers List: {numbers}")

# A list with mixed data types
mixed_list = ["apple", 10, True, 3.14]
print(f"Mixed List: {mixed_list}")

# A list of lists (nested list)
nested_list = [["a", "b"], [1, 2, 3], ["c"]]
print(f"Nested List: {nested_list}")

# Checking the type
print(f"Type of numbers: {type(numbers)}")

# Getting the length
print(f"Length of mixed_list: {len(mixed_list)}")
```

### Indexing

Like strings, lists use **zero-based indexing** to access individual elements. **Negative indexing** accesses elements from the end of the list, where $-1$ is the last element.

```python
my_list = ["P", "y", "t", "h", "o", "n"]

# Accessing elements by positive index
print(f"Element at index 0: {my_list[0]}")
print(f"Element at index 3: {my_list[3]}")

# Accessing elements by negative index
print(f"Last element: {my_list[-1]}")
print(f"Second to last element: {my_list[-2]}")
```

### Slicing

**Slicing** allows you to retrieve a range of elements from the list, returning a **new list**. The syntax is `[start:stop:step]`. The `stop` index is **exclusive**.

```python
alphabet = ["a", "b", "c", "d", "e", "f", "g"]

# Slice from index 2 up to (but not including) index 5
slice_1 = alphabet[2:5]
print(f"Slice [2:5]: {slice_1}")

# Slice from the start up to index 3
slice_2 = alphabet[:3]
print(f"Slice [:3]: {slice_2}")

# Slice from index 4 to the end
slice_3 = alphabet[4:]
print(f"Slice [4:]: {slice_3}")

# Slice with a step of 2 (every second element)
slice_4 = alphabet[::2]
print(f"Slice [::2]: {slice_4}")

# Reversing a list using slicing
reversed_list = alphabet[::-1]
print(f"Reversed list: {reversed_list}")
```

### Modifying a List

Since lists are **mutable**, their elements can be changed after creation.

#### Updating Elements

You can change an element by assigning a new value to its index.

```python
colors = ["red", "green", "blue", "yellow"]
print(f"Original list: {colors}")

# Change the element at index 1
colors[1] = "orange"
print(f"Modified list: {colors}")

# Change a range of elements using slicing
colors[2:4] = ["purple", "pink"]
print(f"Modified range: {colors}")
```

#### Appending Items

Items can be added to the end of a list using the **`.append()`** method.

```python
fruits = ["apple", "banana"]
print(f"Before append: {fruits}")

# Append a single item
fruits.append("cherry")
print(f"After append: {fruits}")

# Use .insert(index, item) to add an item at a specific position
fruits.insert(1, "grape")
print(f"After insert: {fruits}")
```

#### Copy vs View

When working with mutable data types like lists, it's crucial to understand the difference between creating a **reference (view)** and creating a **copy**.

  * **View (Reference):** Simple assignment (`list_b = list_a`) creates a reference. Both variables point to the **same list object** in memory. Changing one affects the other.
  * **Copy:** A shallow copy creates a **new list object** but copies the contents. Changing the new list does **not** affect the original. This is done using the slicing technique `[:]` or the built-in function `list()`.

<!-- end list -->

```python
original = [1, 2, 3]

# 1. View (Reference)
view_list = original
view_list[0] = 99  # Change view_list
print(f"Original (after view change): {original}")

# 2. Copy (using slicing or list())
original = [1, 2, 3] # Reset original list
copy_list_1 = original[:] # Using slicing
copy_list_2 = list(original) # Using list() function

copy_list_1[0] = 55 # Change copy_list_1
print(f"Original (after copy change): {original}") # ORIGINAL REMAINS UNCHANGED
```

#### Concatenation Lists

You can combine two or more lists using the **`+`** operator or the **`.extend()`** method.

  * **`+` Operator:** Creates a **new list** by concatenating the operands.
  * **`.extend()` Method:** Appends elements from one list to the end of an **existing list** (in-place modification).

<!-- end list -->

```python
list_1 = [1, 2, 3]
list_2 = [4, 5, 6]

# Concatenation using + operator (creates a NEW list)
combined_list_new = list_1 + list_2
print(f"Concatenated List (+): {combined_list_new}")

# Extend method (modifies list_1 in place)
list_3 = ['a', 'b']
list_4 = ['c', 'd']
list_3.extend(list_4)
print(f"List 3 after extend: {list_3}")
```

### List Methods

Python provides several built-in methods for working with lists.

| Method | Description | Example |
| :--- | :--- | :--- |
| **`.append(item)`** | Adds an item to the end of the list. | `[1].append(2)` $\rightarrow$ `[1, 2]` |
| **`.insert(i, item)`**| Inserts an item at a specified index $i$. | `[1, 3].insert(1, 2)` $\rightarrow$ `[1, 2, 3]` |
| **`.pop(i)`** | Removes and returns the element at index $i$ (defaults to the last item). | `[1, 2, 3].pop(1)` $\rightarrow$ $2$; list is `[1, 3]` |
| **`.remove(item)`** | Removes the first occurrence of the specified item. | `['a', 'b', 'a'].remove('a')` $\rightarrow$ list is `['b', 'a']` |
| **`.sort()`** | Sorts the list in-place (mutates the original list). | `[3, 1, 2].sort()` $\rightarrow$ `[1, 2, 3]` |
| **`.reverse()`** | Reverses the elements of the list in-place. | `[1, 2, 3].reverse()` $\rightarrow$ `[3, 2, 1]` |
| **`.count(item)`** | Returns the number of times an item appears in the list. | `[1, 2, 1].count(1)` $\rightarrow$ $2$ |
| **`.index(item)`** | Returns the index of the first item found. | `[10, 20].index(20)` $\rightarrow$ $1$ |

```python
data = [5, 2, 8, 2, 1]

# Sort (in-place)
data.sort()
print(f"Sorted: {data}")

# Pop (remove and return the last item)
last_item = data.pop()
print(f"Pop item: {last_item}")
print(f"After pop: {data}")

# Remove (removes the first '2')
data.remove(2)
print(f"After remove: {data}")

# Count
count_two = data.count(2)
print(f"Count of 2: {count_two}")
```

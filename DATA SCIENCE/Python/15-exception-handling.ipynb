# Introduction

When you write Python programs, things don't always go as planned. Sometimes your code has mistakes, or unexpected situations arise during execution. This is where understanding how to handle errors/exceptions becomes crucial.

An "error" is a broad term for anything that prevents your program from running correctly or producing the expected output. In Python, errors are known as exceptions and they can generally be categorized into a few types:

* **Syntax Error**: These are mistakes in the grammar or structure of your Python code. The Python interpreter cannot understand what you're trying to do. Your program won't even start running if it has a syntax error.

* **Logical Error**: These are the trickiest. Your program runs without crashing, and the syntax is perfect, but the logic is flawed, leading to incorrect results. Python doesn't complain about logical errors because it executes your instructions precisely; it just doesn't know what your intentions were.

* **Runtime Error**: These errors occur while your program is running. The syntax is fine, and the initial logic might seem okay, but then an unexpected event happens that Python cannot handle. When a runtime error occurs, Python "raises" an exception. If this exception isn't handled, your program will "crash" and stop executing.

## Why Handle Exceptions?
Handling exceptions is a cornerstone of writing robust and user-friendly software. Here's why it's so important:

* **Prevent Program Crashes**: Without exception handling, a single unexpected situation (like trying to divide by zero) can halt your entire program, leading to a poor user experience.

* **Provide User-Friendly Error Messages**: Instead of cryptic Python tracebacks, you can present clear, helpful messages to the user about what went wrong.

* **Ensure Program Robustness and Reliability**: Well-handled exceptions make your program more resilient to unforeseen inputs or conditions.

* **Perform Cleanup Operations**: Even if an error occurs, you might need to close files, release network connections, or save partial work. Exception handling allows you to guarantee these cleanup steps.

## The `try` and `except` Blocks
The core of exception handling in Python revolves around the `try` and `except` blocks. The `try` block lets you test a block of code for errors. The `except` block lets you handle the error.

The `try` block in the code below will generate an exception because Python doesn't support dividing numbers by zero.

```python
try:
  print(5/0)
except:
  print("An exception occurred")
```


Without the try block, the program will crash and raise an error:

```python
print(5/0)
```

## Handling Specific Exceptions

It's good practice to handle specific types of exceptions. Python has many built-in exception types, and each signifies a different kind of problem. Python comes with a hierarchy of built-in exceptions. Here are some of the most common ones you'll encounter as a beginner:

* `ZeroDivisionError`: Raised when the second operand of a division or modulo operation is zero.

* `ValueError`: Raised when a function receives an argument of the correct type but an inappropriate value.

* `TypeError`: Raised when an operation or function is applied to an object of an inappropriate type.

* `FileNotFoundError`: Raised when a file or directory is requested but doesn't exist.

* `IndexError`: Raised when a sequence (like a list or tuple) index is out of range.

* `KeyError`: Raised when a dictionary key is not found.

* `NameError`: Raised when a local or global name is not found (i.e., a variable or function is used before it's defined).


The code below shows an example of handling a specific exception within a function. Instead of letting the raise a Zero division error, the error is handled and the code provides a user friendly error message.

```python
def perform_division(a, b):
    try:
        result = a / b
        print(f"The result of {a} / {b} is: {result}")
    except ZeroDivisionError:
        print("Error: Division by zero is not allowed.")

perform_division(10, 2)
perform_division(5, 0)
```

## Handling Multiple `except` Blocks

You can have multiple `except` blocks to handle different types of exceptions. The order of `except` blocks matters: Python will check them from top to bottom and execute the first `except` block whose exception type matches the raised exception. It's generally best to list the most specific exceptions first and more general ones later.

```python
def process_numbers():
    try:
        num_str1 = input("Enter the first number: ")
        num_str2 = input("Enter the second number: ")

        num1 = int(num_str1)
        num2 = int(num_str2)

        result = num1 / num2
        print(f"Division result: {result}")

    except ZeroDivisionError:
        print("Specific Error: You cannot divide by zero!")
    except ValueError:
        print("Specific Error: Invalid input. Please enter only whole numbers.")
    except Exception as e: # This is a very general exception, catches anything else
        print(f"An unexpected error occurred: {e}")

process_numbers()
```

* If you enter `10` and `0`, ZeroDivisionError will be caught.

* If you enter `abc` and `5`, ValueError will be caught.

* If any other unexpected error occurs (less common for this simple example), the general `Exception` block will catch it.

## Getting Error Information (`as e`)

You can capture the actual exception object using the as keyword. This object often contains useful information about the error.

```python
try:
    my_list = [1, 2, 3]
    print(my_list[5]) # This will cause an IndexError
except IndexError as e:
    print(f"An index error occurred: {e}")
    print(f"The type of error is: {type(e)}")

try:
    d = {"a": 1}
    print(d["b"]) # This will cause a KeyError
except KeyError as err: # You can use any variable name here, 'err' is also common
    print(f"A key error occurred: {err}")
    print(f"Error details: Key '{err.args[0]}' not found.")
```

## The `else` Block (Optional)

The `else` block in exception handling is executed only if no exception occurs in the `try` block. It's a good place to put code that should run only when the "risky" operation was successful.

```python
def safe_calculation(x, y):
    try:
        result = x / y
    except ZeroDivisionError:
        print("Error: Cannot divide by zero.")
    else:
        # This code runs ONLY if no exception occurred in the try block
        print(f"Calculation successful! Result: {result}")
        print("This message is from the else block.")

print("--- Successful division ---")
safe_calculation(20, 4)

print("\n--- Division by zero ---")
safe_calculation(10, 0)
```

## The `finally` Block (Optional)

The `finally` block is guaranteed to execute, regardless of whether an exception occurred in the `try` block, was caught by an `except` block, or if no exception occurred at all.

Its primary use case is for cleanup operations. This is where you put code that absolutely must run to free up resources, close files, or disconnect from networks, even if something went wrong during the main operation.

```python
try:
    print(8/0)
except:
    print("Something went wrong")
else: # Adding an else to the try block is optional
    print("No exception occurred")
finally:
    print("The 'try except' is finished")
```

## Raising Exceptions

Sometimes, you want to stop the execution of your program and signal that something is wrong, even if Python hasn't detected an error itself. You can do this manually using the `raise` keyword. The syntax is as follows: `raise ExceptionType("Error message")`.

You can raise any built-in exception type, or even custom ones you define.

```python
def set_temperature(temp_celsius):
    if not isinstance(temp_celsius, (int, float)):
        raise TypeError("Temperature must be a number.")
    if temp_celsius < -273.15: # Absolute zero
        raise ValueError("Temperature cannot be below absolute zero (-273.15°C).")
    elif temp_celsius > 1000:
        raise ValueError("Temperature is too high for this system.")
    else:
        print(f"Temperature set to {temp_celsius}°C.")
```

```python
print("--- Valid Temperature ---")
try:
    set_temperature(25.5)
except (TypeError, ValueError) as e:
    print(f"Caught error: {e}")

print("\n--- Invalid Type ---")
try:
    set_temperature("hot")
except (TypeError, ValueError) as e:
    print(f"Caught error: {e}")

print("\n--- Below Absolute Zero ---")
try:
    set_temperature(-300)
except (TypeError, ValueError) as e:
    print(f"Caught error: {e}")

print("\n--- Too High Temperature ---")
try:
    set_temperature(1200)
except (TypeError, ValueError) as e:
    print(f"Caught error: {e}")
```

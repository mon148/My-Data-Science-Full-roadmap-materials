### Python Tuple

A **Tuple** is an ordered, immutable sequence of elements. Tuples are defined by enclosing comma-separated items within **parentheses** (`()`). While parentheses are conventional, they are often optional; the commas are what actually define a tuple.

Tuples are typically used to store related pieces of data that shouldn't change, such as coordinates, dates.

```python
# Creating a simple tuple
my_tuple = (10, 20, 30)
print(f"Simple Tuple: {my_tuple}")
print(f"Type: {type(my_tuple)}")

# Tuple with mixed data types
mixed_tuple = ("Alice", 25, 5.8, True)
print(f"Mixed Tuple: {mixed_tuple}")

# Creating a tuple without parentheses (Tuple Packing)
coords = 3, 4, 5
print(f"Tuple Packing: {coords}")
print(f"Type of coords: {type(coords)}")

# Creating a single-element tuple requires a trailing comma
single_element_tuple = (5,)
print(f"Single Element Tuple: {single_element_tuple}")
print(f"Type: {type(single_element_tuple)}")

# Without the comma, it's just an integer inside parentheses
not_a_tuple = (5)
print(f"Not a Tuple: {not_a_tuple}")
print(f"Type: {type(not_a_tuple)}")
```

### Indexing

Similar to strings and lists, tuple elements are accessed using **zero-based indexing** with square brackets (`[]`). **Negative indexing** can also be used to access elements from the end.

```python
weekday_tuple = ("Mon", "Tue", "Wed", "Thu", "Fri")

# Accessing elements by positive index
print(f"Element at index 0: {weekday_tuple[0]}") # Output: Mon
print(f"Element at index 3: {weekday_tuple[3]}") # Output: Thu

# Accessing elements by negative index
print(f"Last element: {weekday_tuple[-1]}")   # Output: Fri
print(f"Second to last element: {weekday_tuple[-2]}") # Output: Thu

# Accessing elements in a nested tuple
nested_tuple = (1, 2, ("a", "b"))
print(f"Element in nested tuple: {nested_tuple[2][0]}") # Output: a
```

### Slicing

**Slicing** is used to extract a subset of elements, which results in a **new tuple**. The syntax is the same as for lists and strings: `[start:stop:step]`.

```python
data_tuple = (100, 200, 300, 400, 500, 600)

# Slice from index 1 up to (but not including) index 4
slice_1 = data_tuple[1:4]
print(f"Slice [1:4]: {slice_1}") # Output: (200, 300, 400)

# Slice from the start up to index 3
slice_2 = data_tuple[:3]
print(f"Slice [:3]: {slice_2}") # Output: (100, 200, 300)

# Slice with a step of 2
slice_3 = data_tuple[::2]
print(f"Slice [::2]: {slice_3}") # Output: (100, 300, 500)

# Reversing a tuple
reversed_tuple = data_tuple[::-1]
print(f"Reversed tuple: {reversed_tuple}") # Output: (600, 500, 400, 300, 200, 100)
```

### Modifying a Tuple

Tuples are **immutable**, which is their defining characteristic. Once created, you **cannot** change, add, or remove items directly within the tuple.

#### Immutability Demonstration

Attempting to change an element will result in a `TypeError`.

```python
immutable_tuple = (1, 2, 3)

immutable_tuple[0] = 99  # Attempt to change the first element
```

#### Workarounds (Indirect Modification)

To "modify" a tuple, you must create a **new tuple**. This is typically done through **type conversion** or **concatenation**.

1.  **Type Conversion:** Convert the tuple to a list, modify the list, and then convert it back to a tuple.

```python
original_tuple = ("apple", "banana", "cherry")

# 1. Convert to a list
temp_list = list(original_tuple)

# 2. Modify the list
temp_list.append("orange")
temp_list[0] = "grape"

# 3. Convert back to a new tuple
new_tuple = tuple(temp_list)

print(f"Original Tuple: {original_tuple}")
print(f"New Tuple: {new_tuple}")
```

2.  **Concatenation:** Use the `+` operator to combine existing tuples or add new elements (must be done as a tuple).

```python
t1 = (1, 2)
t2 = (3, 4)

# Combining two tuples
combined_tuple = t1 + t2
print(f"Combined Tuple: {combined_tuple}") # Output: (1, 2, 3, 4)

# "Adding" a single element (must be a tuple with a comma)
added_element = combined_tuple + (5,)
print(f"Added Element: {added_element}") # Output: (1, 2, 3, 4, 5)
```

### Tuple Methods

Due to their immutability, tuples have far fewer built-in methods than lists. The two most common methods are for locating and counting elements.

| Method | Description | Example |
| :--- | :--- | :--- |
| **`.count(item)`** | Returns the number of times a specified item appears in the tuple. | `(1, 2, 1).count(1)` $\rightarrow$ $2$ |
| **`.index(item)`** | Returns the index of the first occurrence of the specified item. Raises an error if the item is not found. | `('a', 'b', 'c').index('b')` $\rightarrow$ $1$ |

```python
example_tuple = (10, 20, 30, 10, 40, 50)

# Using .count()
count_of_10 = example_tuple.count(10)
print(f"Count of 10: {count_of_10}") # Output: 2

# Using .index()
index_of_30 = example_tuple.index(30)
print(f"Index of 30: {index_of_30}") # Output: 2

# Finding an item that doesn't exist raises a ValueError
example_tuple.index(99)
```

#### Unpacking

A common feature used with tuples is **unpacking**, which allows you to assign elements of a tuple to multiple variables simultaneously.

```python
# Tuple packing
person = ("Jane Doe", 35, "Engineer")

# Tuple unpacking
name, age, job = person

print(f"Name: {name}, Age: {age}, Job: {job}")
```

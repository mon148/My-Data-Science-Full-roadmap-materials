### Python Sets

A **Set** is an unordered collection of unique, immutable elements. Sets are primarily used to perform mathematical set operations (like union, intersection) and to efficiently check for membership.

Sets are defined using **curly braces** (`{}`) or the built-in function **`set()`**.

**Key Characteristics:**

1.  **Unordered:** Items do not have a defined order and cannot be accessed by index.
2.  **Unique:** Duplicate elements are automatically removed.
3.  **Mutable:** You can add or remove items from a set.
4.  **Immutable Elements:** The individual elements within a set must be of an immutable type (e.g., numbers, strings, tuples), but **not** lists or dictionaries.

```python
# Creating a set using curly braces
fruits = {"apple", "banana", "cherry"}
print(f"Fruits Set: {fruits}")
print(f"Type: {type(fruits)}")

# Sets automatically remove duplicates
duplicate_set = {"a", "b", "a", "c", "b"}
print(f"Set with duplicates removed: {duplicate_set}") # Order is not guaranteed

# Creating an empty set (MUST use set(), not {})
empty_set = set()
print(f"Empty Set: {empty_set}")
# Note: {} creates an empty dictionary, not an empty set.
```

### Indexing & Slicing

Since sets are **unordered**, they do **not** support indexing or slicing. Attempting to access an element by its position will result in a `TypeError`.

To access or check for an element, you must use a loop or the membership operator (`in`).

```python
colors = {"red", "green", "blue"}

# Attempting to index a set will fail
print(colors[0])
```

```python
# Checking for membership is the primary way to check for elements
is_blue_present = "blue" in colors
print(f"'blue' is in the set: {is_blue_present}")
```

### Modifying a Set

Sets are **mutable**, meaning you can add or remove items after creation.

#### Adding Items

  * **`.add(item)`:** Adds a single element to the set.
  * **`.update(iterable)`:** Adds multiple elements from any iterable (like a list or another set).

<!-- end list -->

```python
inventory = {"hammer", "nails"}
print(f"Original Inventory: {inventory}")

# Adding a single item
inventory.add("screwdriver")
print(f"After add(): {inventory}")

# Adding multiple items from a list
new_tools = ["wrench", "saw", "nails"] # 'nails' is a duplicate, so it won't be added
inventory.update(new_tools)
print(f"After update(): {inventory}")
```

#### Removing Items

  * **`.remove(item)`:** Removes the specified element. **Raises a `KeyError` if the item is not found.**
  * **`.discard(item)`:** Removes the specified element. **Does NOT raise an error if the item is not found.**
  * **`.pop()`:** Removes and returns an **arbitrary** element. Since sets are unordered, you don't know which item will be removed.
  * **`.clear()`:** Removes all elements from the set.

```python
data = {1, 2, 3, 4, 5}

# Using .remove()
data.remove(3)
print(f"After remove(3): {data}")

# Using .discard()
data.discard(99) # 99 is not there, no error
print(f"After discard(99): {data}")

# Using .pop() (the element removed is arbitrary)
removed_item = data.pop()
print(f"Popped item: {removed_item}")
print(f"After pop(): {data}")
```

### Set Operations

We can perform Set operations such as, union and intersection from mathematics. These output of this operations return a **new set**.

```python
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

# Union (A or B or both)
union_set = A | B
print(f"Union (A | B): {union_set}") # Output: {1, 2, 3, 4, 5, 6}

# Intersection (A and B)
intersection_set = A.intersection(B) # Or A & B
print(f"Intersection (A & B): {intersection_set}") # Output: {3, 4}

# Difference (A but not B)
difference_set = A - B
print(f"Difference (A - B): {difference_set}") # Output: {1, 2}

# Symmetric Difference (A or B, but not both)
sym_diff_set = A ^ B
print(f"Symmetric Difference (A ^ B): {sym_diff_set}") # Output: {1, 2, 5, 6}
```

### Set Methods

Besides the operational methods above, other useful methods include comparison and copying.

  * **`.issubset(other)`:** Returns `True` if all items in the current set are present in the other set.
  * **`.issuperset(other)`:** Returns `True` if all items in the other set are present in the current set.
  * **`.isdisjoint(other)`:** Returns `True` if the two sets have **no items in common** (their intersection is empty).

<!-- end list -->

```python
set_x = {1, 2}
set_y = {1, 2, 3, 4}
set_z = {5, 6}

# issubset
print(f"Is X a subset of Y? {set_x.issubset(set_y)}")   # True

# issuperset
print(f"Is Y a superset of X? {set_y.issuperset(set_x)}") # True

# isdisjoint
print(f"Are X and Z disjoint? {set_x.isdisjoint(set_z)}") # True (no common elements)
```

### Frozen Set

A **frozenset** is an **immutable version of a set**. Once created, items cannot be added or removed.

**Why use frozenset?** Because frozensets are hashable (immutable), they can be used as **keys in a dictionary** or as **elements within another set**, which regular (mutable) sets cannot.

```python
# Create a regular set
regular_set = {1, 2, 3}

# Create a frozenset
f_set = frozenset([10, 20, 30])
print(f"Frozenset: {f_set}")

# Attempting to add will fail
f_set.add(40)
```

```python
# Frozenset can be used as a dictionary key
my_dict = {f_set: "Immutable Set Key"}
print(f"Dictionary access: {my_dict[f_set]}")
```

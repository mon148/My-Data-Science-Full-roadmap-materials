### Variables

A **variable** is a named storage location that holds a value. In Python, you don't need to declare the type of a variable; you simply assign a value to it, and Python figures out the type automatically.

#### Rules for Variable Names:

1.  Must start with a letter (a-z, A-Z) or an underscore (`_`).
2.  Cannot start with a number.
3.  Can only contain alphanumeric characters and underscores (`A-z, 0-9, and _`).
4.  Names are **case-sensitive** (`age`, `Age`, and `AGE` are three different variables).
5.  Cannot use reserved keywords (like `print`, `for`, `if`, etc.).

```python
# Variable assignment
name = "Charlie"
age = 30
is_active = True
annual_salary = 65000.50

# You can reassign variables
age = 31

# Assigning multiple values in one line
x, y, z = 10, 20, 30

print(f"Name: {name}, Age: {age}")
print(x + y + z)
```

### Displaying Values

The **`print()`** function is the primary way to output data to the console.

You can print multiple values by separating them with a comma. By default, `print()` adds a space between the values and a newline character at the end.

```python
item = "Laptop"
price = 1200

# Basic printing
print("The price is:", price)

# Using an f-string (formatted string literal) for clean output (preferred method)
print(f"The {item} costs ${price}.")

# Changing the separator (sep) and ending character (end)
print("A", "B", "C", sep='-', end='!\n')
print("Next line.")
```

### Data Types

Python has several built-in data types. These are broadly categorized into fundamental types and collection types. The **`type()`** function is used to check the data type of any variable.

#### Fundamental Types

| Category | Type | Description | Example |
| :--- | :--- | :--- | :--- |
| **Numeric** | `int` | Integers (whole numbers). | `10`, `-500` |
| | `float` | Floating point numbers (with decimal part). | `3.14`, `1.0` |
| **Text** | `str` | Strings (sequence of characters). | `"Hello"`, `'Python'` |
| **Boolean** | `bool` | Boolean (True or False). | `True`, `False` |

#### Collection Types

These types are designed to hold multiple items (collections of data).

| Type | Description | Key Characteristic | Example |
| :--- | :--- | :--- | :--- |
| **`list`** | Ordered, changeable (mutable) collection. | Defined by square brackets `[]`. | `["apple", "banana"]` |
| **`tuple`** | Ordered, unchangeable (immutable) collection. | Defined by parentheses `()`. | `(10, 20, 30)` |
| **`set`** | Unordered, unindexed collection of **unique** items. | Defined by curly braces `{}`. | `{"red", "green", "blue"}` |
| **`dict`** | Unordered collection of **key-value pairs**. | Defined by curly braces `{}` with colons `:`. | `{"name": "Jane", "age": 25}` |

```python
a = 10         # int
b = 5.5        # float
e = ["a", "b", "c"] # list
f = ("x", "y") # tuple
g = {1, 2, 3}  # set
h = {"k": 1}   # dict

print(f"Type of list ({e}): {type(e)}")
print(f"Type of tuple ({f}): {type(f)}")
print(f"Type of set ({g}): {type(g)}")
print(f"Type of dictionary ({h}): {type(h)}")
```

### Type Casting

**Type casting** is the process of converting a variable from one data type to another using functions like `int()`, `float()`, `str()`, `list()`, `tuple()`, and `set()`.

```python
num_str = "100"
my_list = [1, 2, 3]

# String to Integer
new_int = int(num_str)
print(f"String to Int: {new_int} (Type: {type(new_int)})")

# List to Tuple
new_tuple = tuple(my_list)
print(f"List to Tuple: {new_tuple} (Type: {type(new_tuple)})")

# List to Set (removes duplicates if they existed)
new_set = set([1, 1, 2, 3])
print(f"List to Set: {new_set} (Type: {type(new_set)})")
```

### Arithmetic Operators

These operators perform mathematical calculations.

| Operator | Name | Example | Result |
| :--- | :--- | :--- | :--- |
| **`+`** | Addition | `5 + 2` | `7` |
| **`-`** | Subtraction | `5 - 2` | `3` |
| **`*`** | Multiplication | `5 * 2` | `10` |
| **`/`** | Division (Float) | `5 / 2` | `2.5` |
| **`//`** | Floor Division (Integer) | `5 // 2` | `2` |
| **`%`** | Modulus (Remainder) | `5 % 2` | `1` |
| **`**`** | Exponentiation | `5 ** 2` | `25` |

```python
x = 10
y = 3

print(f"Addition (10 + 3): {x + y}")
print(f"Float Division (10 / 3): {x / y}")
print(f"Floor Division (10 // 3): {x // y}")
```

### Comparison Operators

These operators compare two values and return a Boolean result (`True` or `False`).

| Operator | Name | Example |
| :--- | :--- | :--- |
| **`==`** | Equal to | `x == y` |
| **`!=`** | Not equal to | `x != y` |
| **`>`** | Greater than | `x > y` |
| **`<`** | Less than | `x < y` |
| **`>=`** | Greater than or equal to | `x >= y` |
| **`<=`** | Less than or equal to | `x <= y` |

```python
a = 15
b = 10

print(f"a == b: {a == b}")
print(f"a > b: {a > b}")
print(f"a != b: {a != b}")
```

### Logical Operators

These operators combine conditional statements and return a Boolean result.

| Operator | Description | Example |
| :--- | :--- | :--- |
| **`and`** | Returns `True` if both statements are true. | `x > 5 and x < 10` |
| **`or`** | Returns `True` if at least one of the statements is true. | `x > 5 or x < 4` |
| **`not`** | Reverses the result; returns `False` if the result is true. | `not (x > 5)` |

```python
temperature = 25
is_weekend = True

# Check if it's warm AND it's the weekend
condition_1 = temperature > 20 and is_weekend
print(f"Is it warm and weekend? {condition_1}")

# Use of 'not'
condition_3 = not is_weekend
print(f"Is it NOT the weekend? {condition_3}")
```

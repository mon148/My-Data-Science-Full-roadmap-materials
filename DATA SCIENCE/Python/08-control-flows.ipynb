## Control Flow

Control flow statements determine the order in which code is executed. Python provides several control flow structures:

**Conditional Statements (`if`, `elif`, `else`):** Used to execute different blocks of code based on whether a condition is true or false.



```python
score = 60
if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"
print(f"Grade: {grade}")
```

```python
fruits = ["apple", "banana"]

if "apple" in fruits:
    print("Apple is in the list.")
else:
    print("Go check am")
```

```python
fruits = ["mango", "banana"]

if "apple" in fruits:
    print("Apple is in the fruits list.")
else:
    print("Apple is not in the fruits list.")
```

**Looping Statements (`for`, `while`):** Used to execute a block of code repeatedly.

**`for` loop:** Iterates over a sequence (like a list, tuple, or string) or other iterable objects.

```python
# Multiplying the element in a list by 2
numbers = [1, 2, 3, 4, 5]
for num in numbers:
    print(num * 2)
```

This is what the `for` loop does in the background.

```python
num = numbers[0] * 2
print(num)
num = numbers[1] * 2
print(num)
num = numbers[2] * 2
print(num)
num = numbers[3] * 2
print(num)
num = numbers[4] * 2
print(num)
```

```python
# Multiplying the element in a list by 2 and store in a new list
numbers = [1, 2, 3, 4, 5]
multiples_of_2 = []

for num in numbers:
    result = num * 2
    multiples_of_2.append(result)
```

**`while` loop:** Executes a block of code as long as a condition is true.

```python
count = 0
while count < 5:
    print(f"Count is: {count}")
    count += 1 # or count = count + 1
```

```python
numbers = [1, 2, 3, 4, 5]
```

```python
i = 0
while i < len(numbers):
    result = numbers[i] * 2
    print(result)
    i += 1
```

* **`break` and `continue` Statements:**
    * **`break`:** Immediately terminates the innermost loop.
    
    * **`continue`:** Skips the rest of the current iteration of the loop and proceeds to the next iteration.

* **`pass` Statement:** A null operation. It is used when a statement is syntactically required but you don't want to execute any code.

```python
numbers = [1, 2, 3, 4]
```

```python
for num in numbers:
    pass
```

```python
# Example of break statement with for loop

for num in numbers:
    if num == 3:
        break
    print(num)
```

```python
# Example of continue statement with for loop

for num in numbers:
    if num == 3:
        print("Skipping '3' going to the next item")
        continue
    print(num)
```

```python
num = 1

while True:
    if num == 10:
        break
    print(num * 2)
    num += 1
```

```python
num = 1

while num < 10:
    print(num * 2)
    num += 1

    if num == 5:
        continue
```

### Inner loops

We can also have a `for` or `while` loop inside another `for` or `while` loop.

```python
my_list = [1, 2]
my_list2 = ["a", "b"]

for i in my_list:
    print("my_list item:", i)
    for j in my_list2:
        print("my_list2 item", j)
```

```python
my_list = [1, 2]
my_list2 = ["a", "b"]

i = 0
while i < len(my_list):
    print("="*10 + "Outer Loop Starts" + "="*10)
    print("my_list item:", my_list[i])
    j = 0
    while j < len(my_list2):
        print("-"*10 + "Inner Loop Starts" + "-"*10)
        print("my_list2 item", my_list2[j])
        j += 1
        print("-"*10 + "Inner Loop Ends" + "-"*10)
    i+=1
    print("="*10 + "Outer Loop Ends" + "="*10)
```

We can also have a `for` loop inside a `while` loop. Or a while loop inside a for loop. 

```python
my_list = [1, 2]
my_list2 = ["a", "b"]

i = 0
while i < len(my_list):
    print("my_list item:", my_list[i])
    j = 0
    for j in my_list2:
        print("my_list2 item", j)
    i+=1
```

Let's look at an example.

```python
numbers_1 = [1, 2, 3, 4, 5]
numbers_2 = [1, 2, 3, 4, 5]

for num1 in numbers_1:
    multiples = []
    for num2 in numbers_2:
        multiples.append(num1 * num2)
    print(f"Multiples of {num1}:", multiples)
```

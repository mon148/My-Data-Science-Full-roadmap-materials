## Modules

Modules are files containing Python definitions and statements. They allow you to organize your code into reusable units.

## Importing Modules
Use the `import` keyword to bring modules into your current script.

* `import module_name`: Imports the entire module. You access its contents using `module_name.attribute` or `module_name.function()`.

* `from module_name import specific_item`: Imports specific functions or attributes directly. You can then use them without the module name prefix.

* `import module_name as alias`: Imports the module with a shorter alias.

````python
import math  # Import the entire math module
print(math.sqrt(16))
````

### Creating Your Own Modules

Creating your own modules in Python is straightforward. You simply save Python code in a file with a `.py` extension. This file then becomes a module that can be imported into other Python scripts.

**Steps to Create a Module:**

1.  **Create a Python file:** Use a text editor or an Integrated Development Environment (IDE) to create a new file. Give it a descriptive name that reflects its purpose, and ensure it ends with the `.py` extension (e.g., `my_math_operations.py`, `data_processing_utils.py`).

2.  **Write Python code in the file:** Define functions, classes, variables, or any other valid Python code that you want to be part of your module.

    **Example: `my_math_operations.py`**
    ```python
    def add(x, y):
        """This function adds two numbers."""
        return x + y

    def multiply(x, y):
        """This function multiplies two numbers."""
        return x * y

    PI = 3.14159

    def circle_area(radius):
        """Calculates the area of a circle."""
        return PI * radius ** 2

    if __name__ == "__main__":
        # This block of code will only run when the script is executed directly
        print("This is the my_math_operations module.")
        print(f"Example addition: {add(5, 3)}")
        print(f"Value of PI: {PI}")
    ```

    The `if __name__ == "__main__":` block is a common practice. Code inside this block will only execute when the module file is run directly (e.g., `python my_math_operations.py`). When the module is imported into another script, this block is skipped. This allows you to include test code or examples within your module without them running when the module is imported.

### Importing Your Own Modules

Once you have created your module file, you can import it into other Python scripts and notebooks using the `import` statement, just like you import built-in or third-party modules. Python needs to be able to find your module file. It typically looks in the following locations:

* The directory where the script being run is located.
* Installation-dependent default paths.

For simple projects, keeping your module file in the same directory as your main script is usually sufficient.

**Examples of Importing Your Custom Module:**

1.  **Importing the entire module:**

````python
import my_math_operations

result_add = my_math_operations.add(10, 5)
result_multiply = my_math_operations.multiply(4, 6)
area = my_math_operations.circle_area(7)
print(f"Addition result: {result_add}")
print(f"Multiplication result: {result_multiply}")
print(f"Circle area: {area}")
print(f"Value of PI from module: {my_math_operations.PI}")
````

2.  **Importing specific items from the module:**

````python
from my_math_operations import add, PI

result = add(20, 7)
print(f"Addition result: {result}")
print(f"Value of PI: {PI}")

# You can also import everything using the asterisk (not generally recommended for large modules)
# from my_math_operations import *
# result_mult = multiply(2, 8)
````


3.  **Importing the module with an alias:**

````python
import my_math_operations as mmo

product = mmo.multiply(3, 9)
print(f"Product: {product}")
print(f"PI using alias: {mmo.PI}")
````

# Introduction

Textual analysis is the process of extracting meaningful information from text data using computational techniques. Text data refers to information primarily composed of human language, expressed as written words, sentences, and paragraphs. Textual analysis combines natural language processing (NLP), machine learning, and statistical analysis to convert large volumes of raw text (such as social media posts, customer reviews, emails, or documents) into structured, actionable insights.

## Challenges of Working With Text Data

While humans effortlessly understand language, computers struggle due to several factors:

* **Ambiguity and Context Understanding**: Many words have multiple meanings , and understanding the correct sense depends on context. Detecting sarcasm, irony, idioms, and subtle sentiments is challenging for algorithms, often leading to misinterpretations

* **Data Quality and Noise**: Text often contains spelling errors, grammatical mistakes, slang, jargon, emojis, and inconsistent writing styles, which can significantly reduce analysis accuracy. Handling multilingual content and irrelevant information further complicates quality management.

* **Handling Rare Words and Sparse Data**: Rare terms or phrases may lack sufficient examples for reliable analysis, posing challenges for sentiment or topic detection.

* **Bias and Representativeness**: If training data is biased or unrepresentative, models built on it may produce inaccurate or unfair results, limiting the generalizability of insights.

## Application of Texual Analysis

Despite the challenges, textual data often contain useful information. By analyzing text, we can unlock insights that power numerous business decisions. Here are three application of textual analysis among many:

* **Sentiment Analysis**: This involves determining the emotional tone behind a body of text, such as customer reviews or social media posts. It helps businesses understand public opinion, customer satisfaction, and brand perception by classifying text as positive, negative, or neutral.

* **Social Media Monitoring**: Textual analysis is used to track mentions of a company, product, or service across social media platforms. It enables organizations to gauge public sentiment, address customer concerns promptly, and manage their online reputation effectively.

* **Product Development and Customer Feedback Analysis**: By analyzing customer reviews, surveys, and feedback, companies can identify pain points, preferences, and feature requests. This insight guides product improvements and innovation to better meet customer needs.



# Processing Textual Data

Raw textual data is inherently unstructured and contains many inconsistencies that make it difficult for machine learning models to process directly. ML algorithms work best with numerical data in a structured format. Text preprocessing is the crucial step of transforming raw text into a clean, normalized, and usable format. The goal is to reduce noise, standardize the text, reduce the vocabulary size, and make the text more amenable to numerical representation for machine learning algorithms.

### Tokenization

Tokenization is the process of breaking down a continuous text into smaller units called tokens. Tokens are usually words, but can also be punctuation marks, numbers, or even sub-word units depending on the tokenization strategy.

#### Using `NTLK` for tokenization.

NLTK (Natural Language Toolkit) is a comprehensive, open-source Python library designed for natural language processing (NLP) tasks involving human language data. It provides a wide range of tools and resources for text processing, including tokenization, classification, stemming, tagging, parsing, and semantic reasoning.

#### Word Tokenization: Splits text into individual words and punctuation


````python
import nltk
````

````python
# nltk.download('punkt_tab')
````

````python
text = "Hello, world! This is a sample sentence."
word_tokens = nltk.word_tokenize(text)
print(f"Word tokens: {word_tokens}")
````

#### Sentence Tokenization: Splits text into individual sentences

````python
text = "First sentence. Second sentence! Is this the last one?"
sent_tokens = nltk.sent_tokenize(text)
print(f"Sentence tokens: {sent_tokens}")
````

For very simple cases, you can use Python's built-in `re` module.

````python
import re
text = "Hello, world! This is a sample sentence."
# This regex splits by non-word characters, keeping only letters and numbers
simple_tokens = re.findall(r'\b\w+\b', text.lower())
print(f"Simple Regex tokens: {simple_tokens}")
````

#### Lower casing & punctuation removal

The goal here is treat "Word", "word", and "WORD" as the same token. This significantly reduces the vocabulary size and ensures consistency.

````python
text = "Hello, World! How are you?"
lower_text = text.lower()
print(f"Lowercased: {lower_text}")

# Apply after tokenization
tokens = ["Hello", ",", "World", "!", "How", "are", "you", "?"]
lower_tokens = [token.lower() for token in tokens]
print(f"Lowercased tokens: {lower_tokens}")
````

Punctuation marks (like '.', ',', '!', '?', ';', ':', '"') often don't carry significant semantic meaning for many ML tasks and can be removed to reduce noise. We can use Python's `string` module and `str.translate()` or regular expressions to remove punctuations.

##### Option 1: Using str.translate

````python
import string

text_with_punct = "Hello, World! This is a sample text with punctuation."

# Create a translation table that maps each punctuation character to None (removal)
translator = str.maketrans('', '', string.punctuation)
text_no_punct_1 = text_with_punct.translate(translator)
print(f"No punctuation (translate): {text_no_punct_1}")
````

##### Option 2: Using re.sub

````python
text_no_punct_2 = re.sub(r'[^\w\s]', '', text_with_punct) # Keep only word characters and spaces
print(f"No punctuation (regex): {text_no_punct_2}")
````

##### Option 3: Keeping characters which are alphanumeric (often combined with tokenization and lowercasing)

````python
tokens = ["Hello", ",", "World", "!", "This", "is", "a", "sample", "text", "with", "punctuation", "."]
cleaned_tokens = [token.lower() for token in tokens if token.isalnum()] # isalnum checks if all chars are alphanumeric
print(f"Cleaned tokens (alphanum check): {cleaned_tokens}")
````

### Stop Word Removal

Stop words are common words in a language that typically carry little unique information and appear very frequently across documents. Examples in English include "the", "a", "is", "and", "of", "to", "in", etc.  Removing stop words can reduce the dimensionality of your feature space (fewer unique words), speed up processing, and sometimes improve model performance by focusing on more meaningful words. While often beneficial, stop word removal is not always appropriate. For tasks like machine translation or sentiment analysis where context or even "not" can be crucial, removing stop words might be detrimental.

#### Using `NLTK`'s stop word list

````python
from nltk.corpus import stopwords
````

````python
# Get the list of English stop words
stop_words = set(stopwords.words('english'))
print(f"First 10 stop words: {list(stop_words)[:10]}")

# Remove stop words from a tokenized text
text_tokens = ["this", "is", "a", "sample", "sentence", "and", "it", "contains", "some", "stop", "words"]
filtered_tokens = [word for word in text_tokens if word not in stop_words]
print(f"Tokens after stop word removal: {filtered_tokens}")
````

### Stemming & Lemmatization

These techniques aim to reduce inflectional forms of words to a common base form, known as the root or lemma. This helps consolidate different forms of the same word (e.g., "running", "ran", "runs" all becoming "run") and further reduces vocabulary size.

#### Stemming

A heuristic process that chops off suffixes from words, often resulting in a "root" word that may not be a true dictionary word. It's faster but less accurate. For example, "running" -> "run", "beautiful" -> "beauti", "history" -> "histori".

`nltk` provides several stemming algorithms, however we'll use the widely used algorithm for English stemming (Porter Stemmer)

````python
from nltk.stem import PorterStemmer

porter_stemmer = PorterStemmer()

words = ["running", "runs", "ran", "runner", "reduce", "reduction", "reducing", "easily", "beautiful", "histories"]

print("--- Porter Stemmer ---")
for word in words:
    print(f"{word} -> {porter_stemmer.stem(word)}")
````

Stemming is aggressive and can result in non-dictionary words. "beautiful" -> "beauti" is a classic example. This might be fine for some statistical models but less ideal if you need human readability or precise word meanings.

### Lemmatization

A more sophisticated and linguistically informed process that reduces words to their base or dictionary form (the lemma). It uses vocabulary and morphological analysis (word structure) to ensure the base form is a valid word. It's slower but more accurate. For example, "running" -> "run", "better" -> "good", "am" -> "be". It requires lexical knowledge base (like WordNet).

````python
from nltk.stem import WordNetLemmatizer
# nltk.download('wordnet') # Uncomment and run this line once

lemmatizer = WordNetLemmatizer()

words = ["running", "runs", "ran", "better", "geese", "mice", "am", "are", "is"]

print("\n--- WordNet Lemmatizer (Default: Noun) ---")
for word in words:
    print(f"{word} -> {lemmatizer.lemmatize(word)}")
# Output examples: running -> running (default is noun), better -> better (default is noun)
# This shows the default behavior. For verbs/adjectives, we need to specify POS tag.

print("\n--- WordNet Lemmatizer (with POS Tag) ---")
print(f"running (verb) -> {lemmatizer.lemmatize('running', pos='v')}") # 'v' for verb
print(f"runs (verb) -> {lemmatizer.lemmatize('runs', pos='v')}")
print(f"ran (verb) -> {lemmatizer.lemmatize('ran', pos='v')}")
print(f"better (adj) -> {lemmatizer.lemmatize('better', pos='a')}") # 'a' for adjective
print(f"geese (noun) -> {lemmatizer.lemmatize('geese', pos='n')}") # 'n' for noun (default)
print(f"mice (noun) -> {lemmatizer.lemmatize('mice', pos='n')}")
print(f"am (verb) -> {lemmatizer.lemmatize('am', pos='v')}")
````

Lemmatization is generally preferred when accuracy and valid words are important, as it relies on a dictionary. Stemming is faster and can be sufficient when you only need to reduce words to a common representation and don't care if the result is a real word. For many ML tasks, the slight accuracy gain from lemmatization might not always justify the increased computational cost, so the choice depends on the specific application.

# Exploratory Analysis of Text Data

Just like with numerical data, before jumping into complex machine learning models, it's crucial to perform Exploratory Data Analysis (EDA) on your textual data. EDA helps you understand the characteristics, patterns, and potential issues within your text corpus. Some EDA techniques for text data inclusde frequency distribution of words, length of documents/sentences, checking the unique words vs total words, and plotting word cloud.

## Word Frequency Distribution

Identifying the most common words (and whether stop words are still present). This useful for understanding the general vocabulary of your corpus. You can use `collections.Counter` or `FreqDist` from `NLTK`.

````python
from collections import Counter
import nltk

# Example Preprocessed Text (similar to what you'd get from Module 2)
sample_corpus = [
    "cat sat mat",
    "dog barked cat",
    "cat dog common pets"
]

# Combine all tokens for frequency analysis
all_words = []
for doc in sample_corpus:
    all_words.extend(doc.split()) # Assuming space-separated tokens

# Count word frequencies
word_counts = Counter(all_words)
print("Most Common Words:")
print(word_counts.most_common(10)) # Top 10 most frequent words
````

````python
# You might want to visualize this with a bar chart using matplotlib/seaborn

import matplotlib.pyplot as plt

common_words = word_counts.most_common(10)
words, counts = zip(*common_words)
plt.figure(figsize=(10, 6))
plt.bar(words, counts)
plt.title('Top 10 Most Common Words')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
````

## Length of Document/Sentences

Understanding the average, minimum, and maximum length can inform model design. This also helps us identify very short or very long documents that might need special handling.

````python
document_lengths = [len(doc.split()) for doc in sample_corpus]
print(f"\nDocument Lengths: {document_lengths}")
print(f"Average Document Length: {sum(document_lengths) / len(document_lengths):.2f}")
print(f"Max Document Length: {max(document_lengths)}")
print(f"Min Document Length: {min(document_lengths)}")
````

````python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))=
plt.hist(document_lengths, bins=5, edgecolor='black')
plt.title('Distribution of Document Lengths')
plt.xlabel('Number of Words')
plt.ylabel('Number of Documents')
plt.show()
````

## Unique Words vs. Total Words (Lexical Diversity):

Lexical diversity refers to the range of unique words used in a text.  It is commonly measured by the ratio of unique words (types) to the total number of words (tokens) in a text, a metric known as the Type-Token Ratio (TTR).  A high TTR indicates high lexical diversity. This can be calculated per document or for the entire corpus.

````python
total_words = len(all_words)
unique_words = len(set(all_words))
print(f"\nTotal words in corpus: {total_words}")
print(f"Unique words in corpus: {unique_words}")
print(f"Lexical Diversity (Unique/Total): {unique_words / total_words:.2f}")
````

# Extracting Features from Text

Machine learning algorithms, at their core, operate on numerical data. They understand numbers, not raw words or sentences. Therefore, a critical step after preprocessing is feature extraction (or feature engineering): converting the cleaned textual data into a numerical representation that can be fed into an ML model.

The most common approaches involve creating a vocabulary of all unique words (or "tokens") present in your dataset and then representing each document as a vector where each dimension corresponds to a word in the vocabulary.

## Bag-of-Words (BoW)

The Bag-of-Words (BoW) model is one of the simplest and most common ways to represent text as numerical features. It treats a document (e.g., a sentence, paragraph, or entire article) as an unordered collection ("bag") of words. Each document is represented by a vector, where each element in the vector corresponds to the frequency (count) of a word from the predefined vocabulary in that document.

Here is how to it works:

1. **Create a Vocabulary**: Collect all unique words from your entire corpus (all documents combined). This forms your vocabulary. Each unique word is assigned a unique index.

2. **Vectorize Documents**: For each document, create a vector of the same length as your vocabulary.

3. **Count Occurrences**: For each word in the document, increment the count at its corresponding index in the vector.

#### Implementing BoW with `sklearn.feature_extraction.text.CountVectorizer`:
CountVectorizer handles the entire process: tokenization (if not pre-tokenized), vocabulary creation, and word counting. It's designed to work directly on raw text or preprocessed text.

````python
from sklearn.feature_extraction.text import CountVectorizer

corpus = [
    "This is the first document.",
    "This document is the second document.",
    "And this is the third one.",
    "Is this the first document?"
]

# Initialize CountVectorizer
# We can pass preprocessor and tokenizer functions if we did custom preprocessing
# For simplicity, let's use its built-in tokenizer for this example,
# but in a real pipeline, you'd feed it your preprocessed text.
vectorizer = CountVectorizer()

# Fit the vectorizer to the corpus to learn the vocabulary
vectorizer.fit(corpus)

# Get the vocabulary (word-to-index mapping)
print("Vocabulary:")
print(vectorizer.vocabulary_)
# Note: Indices are assigned automatically by CountVectorizer

# Transform the corpus into BoW feature vectors
X_bow = vectorizer.transform(corpus)

# Print the sparse matrix (often sparse because most documents don't contain all words)
print("\nBoW Sparse Matrix:")
print(X_bow) # This shows (document_index, word_index) count

# Convert to a dense array for easier viewing (for small datasets)
print("\nBoW Dense Array:")
print(X_bow.toarray())

# Get feature names (words) corresponding to columns
print("\nFeature names (words):")
print(vectorizer.get_feature_names_out())
````

#### Limitations of BoW:

* **No Semantic Meaning/Context**: It doesn't capture the meaning or relationship between words. "Dog bites man" and "Man bites dog" would have the same representation, even though their meanings are entirely different.

* **Sparsity**: For large vocabularies and many documents, most word counts will be zero, leading to very sparse matrices, which can be computationally intensive.

* **High Dimensionality**: The feature vector's length is equal to the vocabulary size, which can be millions for very large text datasets.

## Term Frequency-Inverse Document Frequency (TF-IDF)

TF-IDF is a more sophisticated weighting scheme than simple word counts, designed to reflect how important a word is to a document in a collection or corpus. Words that appear very frequently in *one specific document* but rarely in *other documents* are considered more important and given a higher weight. Words that appear frequently across all documents (like stop words) are given lower weights, even if they appear many times in a single document.

#### Components of TF-IDF

* **Term Frequency (TF)**: How often a word appears in a document.

    $$TF(t,d)=Count of term t in document d$$

    (Sometimes normalized by total words in document)
  
* **Inverse Document Frequency (IDF)**: A measure of how rare or common a word is across the entire corpus. Words that appear in many documents have a low IDF, while words that appear in few documents have a high IDF.

    $$
    IDF(t,D)=log\left( \frac{Total number of documents}{Number of documents containing term t}\right)
    $$

    (The log is used to dampen the effect, and a smoothing term +1 is often added to the denominator to avoid division by zero).

* **TF-IDF Score**: The product of TF and IDF.

    $$TFIDF(t,d,D)=TF(t,d)×IDF(t,D)$$

#### When to use TF-IDF vs. BoW:

* TF-IDF is generally preferred when you want to highlight words that are distinctive to a particular document, making it useful for tasks like document classification, information retrieval, or topic analysis where unique terms matter more than common ones.

* BoW can be simpler and sometimes sufficient for tasks where the sheer presence of words is more important, or as a baseline.

#### Implementing TF-IDF with `sklearn.feature_extraction.text.TfidfVectorizer`:
TfidfVectorizer combines the functionality of CountVectorizer with the TF-IDF weighting scheme.

````python
from sklearn.feature_extraction.text import TfidfVectorizer

corpus = [
    "This is the first document.",
    "This document is the second document.",
    "And this is the third one.",
    "Is this the first document?"
]

# Initialize TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer(norm=None)

# Fit and transform the corpus
X_tfidf = tfidf_vectorizer.fit_transform(corpus)

# Print feature names (words)
print("Feature names (words):")
print(tfidf_vectorizer.get_feature_names_out())

# Print the TF-IDF sparse matrix
print("\nTF-IDF Sparse Matrix:")
print(X_tfidf)

# Convert to a dense array for easier viewing
print("\nTF-IDF Dense Array:")
print(X_tfidf.toarray())

# Example: Check IDF for a common word ('the') vs. a less common word ('second')
# You can inspect the idf_ attribute after fitting
print("\nIDF values:")
for word, idx in tfidf_vectorizer.vocabulary_.items():
    if word in ['the', 'second', 'document']: # Check some specific words
         print(f"  Word: '{word}', IDF: {tfidf_vectorizer.idf_[idx]:.4f}")
# Note: 'the' will have a low IDF (common), 'second' will have a higher IDF (rarer)
````

## N-grams

An N-gram is a contiguous sequence of N items from a given sample of text or speech. These "items" can be words, characters, or even syllables. Coomon examples of N-grams include the following:

* **Unigram (1-gram)**: A single word (what BoW and TF-IDF typically use by default).

* **Bigram (2-gram)**: A sequence of two consecutive words (e.g., "quick brown", "brown fox").

* **Trigram (3-gram)**: A sequence of three consecutive words (e.g., "the quick brown", "quick brown fox").

#### Why use N-grams?

* **Capture Local Word Order/Context**: Unlike simple BoW/TF-IDF, N-grams can capture some limited context. "not good" has a different meaning than "good," and a bigram ("not good") can capture this.

* **Phrases**: They can identify common phrases or expressions.

#### Generating N-grams with CountVectorizer and TfidfVectorizer:
Both CountVectorizer and TfidfVectorizer allow you to specify ngram_range to include N-grams as features.

````python
from sklearn.feature_extraction.text import CountVectorizer

corpus = [
    "This is a good movie.",
    "This is not a good movie.",
    "A bad movie.",
    "A good movie is a good movie."
]

# Example with Bigrams (ngram_range=(1, 2) includes unigrams and bigrams)
vectorizer_ngrams = CountVectorizer(ngram_range=(1, 2)) # (min_n, max_n)
X_ngrams = vectorizer_ngrams.fit_transform(corpus)

print("N-gram Features (Unigrams & Bigrams):")
print(vectorizer_ngrams.get_feature_names_out())
# Expected output will include 'good', 'movie', 'not', 'good movie', 'not good', etc.

print("\nN-gram BoW Dense Array:")
print(X_ngrams.toarray())
````

````python
# You can also apply TF-IDF to N-grams
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_ngrams_vectorizer = TfidfVectorizer(ngram_range=(1, 2))
X_tfidf_ngrams = tfidf_ngrams_vectorizer.fit_transform(corpus)
print("\nTF-IDF N-gram features:")
print(tfidf_ngrams_vectorizer.get_feature_names_out())

# Convert to a dense array for easier viewing
print("\nTF-IDF N-gram Dense Array:")
print(X_tfidf_ngrams.toarray())
````

#### The Trade-offs with N-grams:

* **Increased Dimensionality**: As 'N' increases, the number of possible N-grams explodes, leading to a much larger and sparser feature space.

* **Computation**: More memory and processing power are required.

* **Sparsity**: Many N-grams might appear only once or twice, making them less statistically significant.

# Handling Inconsistent Entries

Another thing to take note of our data cleaning stage is inconsistent values in certain columns in our data set.

````python
data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eve", "Frank", "Jane", "John"],
    "City": ["New York", "new york", "NYC", "Los Angeles", "LA", "los angeles", " New York", "losangeles"],
    "Gender": ["Male", "male", "FEMALE", "Female", "female", "MALE", "MaLE", " FeMaLE"],
}

df = pd.DataFrame(data)
df
````

````python
print("Unique values in 'City':", df['City'].unique())
print("Unique values in 'Gender':", df['Gender'].unique())
````

Looking at the data, we can see there are inconsistent values in the "City" and "Gender" column. For example, in the "City" column "New York" and "new york". There are also some values with trailing white spaces in the column.

The first step in fixing this is to standardise the values and remove any white spaces that might be before or after each values.

````python
df["City"] = df["City"].str.lower()
df["Gender"] = df["Gender"].str.lower()
````

````python
print("Unique values in 'City':", df['City'].unique())
print("Unique values in 'Gender':", df['Gender'].unique())
````

All values are now standardised to be in lower case. But there are still values with white spaces. We can fix that using strings "strip" method.

````python
df["City"] = df["City"].str.strip()
df["Gender"] = df["Gender"].str.strip()
````

````python
print("Unique values in 'City':", df['City'].unique())
print("Unique values in 'Gender':", df['Gender'].unique())
````

All inconisistent entries in the gender column has been fixed. However, we still need to do one more step to completely fix the values in the "City" column. For example, "new york" and "nyc". We can fix this by mapping similar value to a consistent label.

````python
city_mapping = {"nyc": "new york", "la": "los angeles", "losangeles": "los angeles"}
df["City"] = df["City"].replace(city_mapping)
````

````python
print("Unique values in 'City':", df['City'].unique())
print("Unique values in 'Gender':", df['Gender'].unique())
````

# Fixing Wrong Data Types

When working with real world data. Columns may have incorrect data types. For example, numbers stored as string, dates stored as plain text, boolean values reprersented as string.

````python
data = {
    "CustomerID": ["001", "002", "003", "004"],
    "Age": ["25", "30", "22", "27"],  # Should be integers
    "SignupDate": [
        "2024/01/10",
        "2024/02/15",
        "2024/03/05",
        "2024/04/20",
    ],  # Should be datetime
    "IsActive": ["True", "False", "TRUE", "false"],  # Should be boolean
}

df = pd.DataFrame(data)
df
````

````python
df.info()
````

````python
df.dtypes
````

````python
# Convert Age to integer
df["Age"] = df["Age"].astype(int)
````

````python
# Convert IsActive to boolean
df["IsActive"] = df["IsActive"].str.lower().map({"true": True, "false": False})
````

````python
# Convert SignupDate to datetime
df["SignupDate"] = pd.to_datetime(df["SignupDate"])
````

````python
df.info()
````

# Parsing Dates

Pandas also allows us to specify the format for parsing dates by providing a string representation of how the date is strucktured. By providing a format string containing special format codes (like %Y for year, %m for month, %d for day), we can customize how the date and time information is represented. For example, "2024.05.08" has the format "%Y.%m.%d"

````python
date_data = {
    "OrderID": [1, 2, 3, 4],
    "OrderDate": ["2024,01,10", "2024,02,15", "2024,03,05", "2024,01,04"]
}

df_dates = pd.DataFrame(date_data)
df_dates
````

````python
pd.to_datetime(df_dates["OrderDate"], format="%Y,%m,%d")
````

````python
df_dates["OrderDateParsed"] = pd.to_datetime(df_dates["OrderDate"], format="%Y,%m,%d")  # Specify format
df_dates
````

There are other possible [parts](https://strftime.org/) we can specify providing the formats for our date string but the most common but the most common are %d for day, %m for month, %y for a two-digit year and %Y for a four digit year.

Dataset sometimes contain different date formats:
- `"2024-01-10"` (ISO format)

- `"2024/02/15"` (slashes)

- `"March 5, 2024"` (long format)

- `"04-01-2024"` (ambiguous: is it April 1 or Jan 4?)

````python
date_data = {
    "OrderID": [1, 2, 3, 4],
    "OrderDate": ["2024-01-10", "2024/02/15", "March 5, 2024", "04-01-2024"]  # Different formats
}

df_dates = pd.DataFrame(date_data)
df_dates
````

````python
# Let pandas automatically infer formats using, assuming January 4 for 04-01-2024
df_dates["OrderDate_parsed"] = pd.to_datetime(df_dates["OrderDate"], format="mixed")

df_dates
````

Pandas automatically converts the dates for us. However, it couldn't the format for "04-01-2024" as January 4.

````python
df_dates["OrderDate_parsed_day_first"] = pd.to_datetime(df_dates["OrderDate"], format="mixed", dayfirst=True)

df_dates
````

If we don't specify a format. Pandas will raise an error for values that does not follow the default format. We can silence this error and replace bad values with missing values by passing `error="coerce"` to the `pd.to_datetime` function.

````python
pd.to_datetime(df_dates["OrderDate"], errors="coerce")
````

# Introduction 
Good data is the backbone of any successful data science or machine learning project. No matter how advanced your model is, if the data feeding into it is low quality or poorly understood, the results will be unreliable. That’s why learning how to explore, understand, and prepare your data is one of the most valuable skills you can develop.

# What is Data?
Data refers to raw facts, figures, symbols, or observations that, on their own, lack context or meaning but can be processed or organized to become useful information.

## Types of data

1. Numerical Data

    **Discrete**: Whole numbers that count things.
    Example: Number of cars in a parking lot.

    **Continuous**: Numbers that can take on any value in a range.
    Example: Temperature readings throughout the day.

2. Categorical Data

    **Nominal**: Categories with no inherent order.
    Example: Colors (Red, Blue, Green).

    **Ordinal**: Categories with a clear, ranked order.
    Example: Education levels (High School < Bachelor's < Master's).

# Structured vs. Unstructured Data
Data can also be classified based on how it's organized:

* **Structured Data**: Highly organized and easy to search, usually stored in tables (like Excel spreadsheets or SQL databases).
    Example: A dataset of customer purchases with columns like name, age, and amount spent.

* **Unstructured Data**: Lacks a predefined structure, making it harder to analyze directly.
    Examples: Emails, social media posts, audio recordings, or images.

In many real-world cases, unstructured data is transformed into structured form — like extracting keywords from text or detecting objects in images — before being used in ML models.

# Data Analysis

Data analysis is the process of examining, organizing, cleaning, and analyzing data to uncover useful information, draw conclusions, and support decision-making.

## The Data Analysis Process

The typical steps in the data analysis process includes the following:

1. **Define the Objective** - Identify the goal of the analysis.

2. **Data Collection** – Gathering data from various sources (CSV files, databases, APIs, etc.).

3. **Data Cleaning** – Fixing or removing incomplete, incorrect, or irrelevant data.

4. **Data Exploration (EDA)** – Using statistics and visualisations to understand patterns, trends, and relationships.

5. **Data Interpretation** – Drawing conclusions and generating insights to inform decisions or build machine learning models.

# Data Analysis in the Machine Learning
In the context of machine learning, here’s a simplified view of a typical Machine Learning (ML) pipeline (structured sequence of steps that transforms raw data into a deployed model capable of making predictions or decisions):

**Data Collection** -> **Data Analysis** -> **Model Training** -> **Model Evaluation** → **Deployment**

Data analysis is crucial in the machine learning workflow because:

* It helps us understand our features (variables) and their distributions.

* It allows us to detect and fix issues like missing values, outliers, or incorrect data types.

* It helps select relevant features and engineer new ones to improve model performance.

* It ensures you’re not training a model on biased, inconsistent, or low-quality data.

Without a solid understanding of our data, the whole ML process can easily go off track.

## Feature Engineering

Feature engineering is the process of using domain knowledge to create new, useful features from raw data. It’s a crucial step in machine learning that can significantly improve model performance. Think of it as preparing and refining the ingredients (raw data) so a chef (the model) can create a better dish.

Feature engineering can be broadly categorized into three forms:

  * **Feature Transformation:** Modifying existing features to improve their representation for the model.
  * **Feature Scaling:** Adjusting the scale of numerical features to a standard range.
  * **Feature Construction:** Creating new features by combining or extracting information from existing ones.

## Feature Transformation

Feature transformation is the process of applying a mathematical function to a feature to make it more suitable for machine learning algorithms. This often includes handling missing values, converting categorical features to numerical values, handling outliers, and scaling features to a common range.

### Handling Missing Values

We can't feed missing values to a machine learning model, so we need to handle them. A simple method is **imputation**, which means filling in the missing values.

#### Example: Imputing with the mean

```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer

# Create a sample DataFrame with missing values
data = {"age": [25, 30, np.nan, 40, 35], "salary": [50000, 60000, 70000, np.nan, 80000]}
X = pd.DataFrame(data)


print("Original DataFrame:")
print(X)

# Use SimpleImputer to fill missing values with the mean
imputer = SimpleImputer(strategy="mean")
imputer.fit(X)
imputed_X = imputer.transform(X)
X_imputed = pd.DataFrame(imputed_X, columns=X.columns)

print("\nDataFrame after mean imputation:")
print(X_imputed)
```

### Handling Categorical Values

Many machine learning models can only work with numerical data. So, we must convert categorical features into numbers. There two common ways of handling categorical values:

* One-hot encoding: For tranforming nominal categories to numeric values.
* Ordinal/label encoding: For transforming ordinal categories to numeric values.

#### Example: One-Hot Encoding

One-Hot Encoding creates a new binary column for each unique category in a feature. A `1` in a column indicates that the row belongs to that category, and a `0` indicates it doesn't.

```python
from sklearn.preprocessing import OneHotEncoder

# Sample data
data = {"city": ["New York", "Paris", "London", "New York", "Paris"]}
X = pd.DataFrame(data)

# Create an instance of OneHotEncoder
encoder = OneHotEncoder(sparse_output=False)

# Fit and transform the data
encoded_features = encoder.fit_transform(X[["city"]])

# Convert the result back to a DataFrame
encoded_X = pd.DataFrame(
    encoded_features, columns=encoder.get_feature_names_out()
)

print("Original DataFrame:")
print(X)
print("\nDataFrame after One-Hot Encoding:")
print(encoded_X)
```

#### Example: Ordinal/Label Encoding

Ordinal encoding creates a new numerical feature from the categorical column. Each category in the feature is mapped to a numeric value, such that the higher the value the more important the category is.

```python
from sklearn.preprocessing import OrdinalEncoder

# Sample data
data = {"city": ["New York", "Paris", "London", "New York", "Paris"]}
X = pd.DataFrame(data)

# Create an instance of OrdinalEncoder
encoder = OrdinalEncoder() # Creates label encoded categories

# Fit and transform the data
encoded_features = encoder.fit_transform(X[["city"]])

# Convert the result back to a DataFrame
encoded_X = pd.DataFrame(
    encoded_features, columns=encoder.get_feature_names_out()
)

print("Original DataFrame:")
print(X)
print("\nDataFrame after One-Hot Encoding:")
print(encoded_X)
```

By default ordinal encoders creates label encoded categories where each categorical mapped to numeric value. However we can specify a custom hierachy by passing our custom list of categories from least important to most important.

```python
# Sample data
data = {"grade": ["Bad", "Excellent", "Good", "Bad", "Excellent"]}
X = pd.DataFrame(data)

# Create an instance of OrdinalEncoder
encoder = OrdinalEncoder(categories=[["Bad", "Good", "Excellent"]]) # Creates label encoded categories

# Fit and transform the data
encoded_features = encoder.fit_transform(X[["grade"]])

# Convert the result back to a DataFrame
encoded_X = pd.DataFrame(
    encoded_features, columns=encoder.get_feature_names_out()
)

print("Original DataFrame:")
print(X)
print("\nDataFrame after One-Hot Encoding:")
print(encoded_X)
```

### Handling Outliers

Outliers are data points that are significantly different from other observations. They can skew the results of a model. Here are three common methods for handling them.

#### 1. Trimming
Trimming (also known as truncation) involves removing the outlier data points entirely. This is the simplest method, but it can lead to a loss of valuable information. This method is best used when you are confident the outliers are due to data entry errors or are not representative of the underlying population.

```python
# Create a DataFrame with an outlier
data = {'age': [20, 25, 30, 35, 120]}  # 120 is an outlier
X = pd.DataFrame(data)

# Calculate the 95th percentile
upper_limit = X['age'].quantile(0.95)

# Trim the data by keeping only values below the upper limit
X_trimmed = X[X['age'] < upper_limit]

print("Original DataFrame with outlier:")
print(X)
print("\nDataFrame after trimming the outlier:")
print(X_trimmed)
```

#### 2. Log Transformation
Log transformation is a powerful technique for handling right-skewed data and reducing the impact of outliers. It compresses the range of the data, especially for large values, making the distribution more symmetrical. It is only applicable to non-negative values.

```python
import numpy as np

# Create a DataFrame with an outlier
data = {'income': [50000, 60000, 70000, 80000, 500000]} # 500,000 is an outlier
X = pd.DataFrame(data)

# Apply a log transformation
X['income_log'] = np.log(X['income'])

print("Original DataFrame with outlier:")
print(X[['income']])
print("\nDataFrame after log transformation:")
print(X)
```

#### 3. Power Transformation
The Power Transformation is a family of transformations that includes the Box-Cox and Yeo-Johnson transformations. It is a powerful method for stabilizing variance and making data more normally distributed. The Box-Cox method requires data to be strictly positive, while the Yeo-Johnson can handle zero and negative values, making it more flexible.

The formula for the Box-Cox transformation $y(\lambda)$ on a data point $y$ is:

$$
y(\lambda) = \begin{cases} \frac{y^{\lambda} - 1}{\lambda} & \text{if } \lambda \neq 0 \\ \ln(y) & \text{if } \lambda = 0 \end{cases}
$$

The formula for the Yeo-Johnson transformation $y(\lambda)$ on a data point $y$ is defined as:

**Case 1: For Non-Negative Values ($y \geq 0$)**

$$
y(\lambda) = \begin{cases} \frac{(y + 1)^{\lambda} - 1}{\lambda} & \text{if } \lambda \neq 0 \\ \ln(y + 1) & \text{if } \lambda = 0 \end{cases}
$$

**Case 2: For Negative Values ($y < 0$)**

$$
y(\lambda) = \begin{cases} - \frac{(-y + 1)^{2 - \lambda} - 1}{2 - \lambda} & \text{if } \lambda \neq 2 \\ - \ln(-y + 1) & \text{if } \lambda = 2 \end{cases}
$$

Where

* $y$ is the original data point.
* $\lambda$ is the transformation parameter, typically optimized to maximize the normality of the transformed data.
* $\ln(y)$ is the natural logarithm of $y$ (or $y+1$ and $-y+1$).

We'll use scikit-learn's PowerTransformer, which automatically determines the best transformation method (Box-Cox or Yeo-Johnson) and finds the optimal parameter to make the data more Gaussian-like.

```python
from sklearn.preprocessing import PowerTransformer

# Create a DataFrame with an outlier and right-skewed data
data = {'price': [100, 200, 300, 400, 2500]} # 2500 is an outlier
X = pd.DataFrame(data)

# Create an instance of PowerTransformer with the 'box-cox' method
# Note: 'yeo-johnson' is the default and can handle negative values
pt = PowerTransformer(method='box-cox')

# Fit and transform the data
X['price_transformed'] = pt.fit_transform(X[['price']])

print("Original DataFrame with outlier:")
print(X[['price']])
print("DataFrame after Power Transformation:")
print(X)
print(f"Fitted lambda value: {pt.lambdas_[0]:.4f}")
```

## Feature Scaling

### What is Feature Scaling?

Feature scaling is a method used to normalize the range of independent variables or features of data. It’s important for algorithms that rely on the distance between data points.

### Standardization

Standardization (or Z-score normalization) scales the features to have a mean of 0 and a standard deviation of 1. The formula is: $$z = (x - \mu) / \sigma$$

```python
from sklearn.preprocessing import StandardScaler

# Sample data
data = {"age": [25, 30, 35, 40], "salary": [50000, 60000, 70000, 80000]}
X = pd.DataFrame(data)

# Standardize the features
scaler = StandardScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)

print("Original DataFrame:")
print(X)
print("\nDataFrame after Standardization:")
print(X_scaled)
```

### Normalization

Normalization (or Min-Max scaling) scales the features to a fixed range, usually 0 to 1. The formula is: $$x_{scaled} = (x - x_{min}) / (x_{max} - x_{min})$$

```python
from sklearn.preprocessing import MinMaxScaler

# Sample data
data = {"age": [25, 30, 35, 40], "salary": [50000, 60000, 70000, 80000]}
X = pd.DataFrame(data)

# Normalize the features
scaler = MinMaxScaler()
X_normalized = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)

print("Original DataFrame:")
print(X)
print("\nDataFrame after Normalization:")
print(X_normalized)
```

## Feature Construction

Feature construction involves creating new features by combining or transforming existing ones. This often requires **domain knowledge**.

### Example: Creating a BMI feature
Let's say we have 'height' and 'weight'. We can construct a new feature, 'BMI', which is more meaningful to a model than the two separate features.

```python
# Sample data
data = {"weight_kg": [70, 80, 65], "height_m": [1.75, 1.80, 1.60]}
X = pd.DataFrame(data)

# Construct a new feature: BMI
X["BMI"] = X["weight_kg"] / (X["height_m"] ** 2)

print("Original DataFrame:")
print(X[["weight_kg", "height_m"]])
print("\nDataFrame with constructed 'BMI' feature:")
print(X)
```

## Using Pipelines for Feature Engineering

A **pipeline** is a sequence of steps that can be chained together. It helps to automate the workflow, prevent data leakage, and ensure that the same transformations are applied to both the training and test datasets.

### Combining Transformation Steps in a Pipeline

We can use `scikit-learn`'s `Pipeline` to chain multiple transformation steps.

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder

# Sample data with numerical and categorical features
data = {
    "age": [25, 30, np.nan, 40],
    "gender": ["male", "female", "female", "male"],
    "salary": [50000, 60000, 70000, 80000],
}
X = pd.DataFrame(data)

# Create a pipeline for numerical features
numerical_pipeline = Pipeline(
    steps=[("imputer", SimpleImputer(strategy="mean")), ("scaler", StandardScaler())]
)

# Create a pipeline for categorical features
categorical_pipeline = Pipeline(
    steps=[
        ("imputer", SimpleImputer(strategy="most_frequent")),
        ("onehot", OneHotEncoder(handle_unknown="ignore")),
    ]
)

# Use ColumnTransformer to apply different pipelines to different columns
preprocessor = ColumnTransformer(
    transformers=[
        ("num", numerical_pipeline, ["age", "salary"]),
        ("cat", categorical_pipeline, ["gender"]),
    ]
)

# Now, we can fit and transform our data
X_processed = preprocessor.fit_transform(X)
X_processed = pd.DataFrame(X_processed, columns=preprocessor.get_feature_names_out())

print("Original DataFrame:")
print(X)
print("\nProcessed data using a pipeline:")
print(X_processed)
```

Let's do the same steps without a pipeline

```python
# Sample data with numerical and categorical features
data = {
    "age": [25, 30, np.nan, 40],
    "gender": ["male", "female", "female", "male"],
    "salary": [50000, 60000, 70000, 80000],
}
X = pd.DataFrame(data)

# Define numerical feature preprocessors
num_imputer = SimpleImputer(strategy="mean")
scaler = StandardScaler()

# Define categorical feature preprocessors
cat_imputer = SimpleImputer(strategy="most_frequent")
ohe = OneHotEncoder(sparse_output=False)

# Preprocess numrical features
X_numerical = num_imputer.fit_transform(X[["age", "salary"]])
X_numerical_scaled = scaler.fit_transform(X_numerical)
X_numerical_scaled = pd.DataFrame(X_numerical_scaled, columns=["age" , "salary"])

# Preprocess categorical features
X_categorical = cat_imputer.fit_transform(X[["gender"]])
X_categorical = ohe.fit_transform(X_categorical)
X_categorical = pd.DataFrame(X_categorical, columns=ohe.categories_[0])

# Combine the output of the preprocessors together
X_processed = pd.concat([X_numerical_scaled, X_categorical], axis=1)

print("Original DataFrame:")
print(X)
print("\nProcessed data using a pipeline:")
print(X_processed)
```

## Building a Custom Scikit-learn Transformer

Sometimes, `scikit-learn` doesn't have a pre-built transformer for a specific task. We can create a **custom transformer** by inheriting from `BaseEstimator` and `TransformerMixin`.

```python
from sklearn.base import BaseEstimator, TransformerMixin


# Custom transformer to create a feature for BMI
class BMICreator(BaseEstimator, TransformerMixin):
    def __init__(self):
        pass

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        # Ensure 'weight_kg' and 'height_m' columns exist
        if "weight_kg" not in X.columns or "height_m" not in X.columns:
            raise ValueError(
                "DataFrame must contain 'weight_kg' and 'height_m' columns."
            )

        # Create the new 'BMI' feature
        X_copy = X.copy()
        X_copy["BMI"] = X_copy["weight_kg"] / (X_copy["height_m"] ** 2)
        return X_copy


# Sample data
data = {"weight_kg": [70, 80, 65], "height_m": [1.75, 1.80, 1.60]}
X = pd.DataFrame(data)

# Use our custom transformer
bmi_transformer = BMICreator()
X_with_bmi = bmi_transformer.fit_transform(X)

print("Original DataFrame:")
print(X)
print("\nDataFrame after applying custom BMICreator transformer:")
print(X_with_bmi)
```
